# Importing libraries 

```python
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC 
from selenium.webdriver.common.action_chains import ActionChains
import pandas as pd
import numpy as np
import requests
import time
```

# Scrapping

```python
url = "https://appsumo.com/browse"
driver = webdriver.Chrome(executable_path=r'C:\Users\SDO\Downloads\chromedriver_win32\chromedriver.exe')
driver.get(url)
html = driver.page_source.encode('utf-8')
page_num = 0

while driver.find_elements_by_link_text('Show More Results'):
    driver.find_element_by_link_text('Show More Results').click()
    page_num += 1
    print("getting page number "+str(page_num))
    time.sleep(1)
    html = driver.page_source.encode('utf-8')
```

```python
soup=BeautifulSoup(html,'html.parser')
```

**data is inside div class : appsumo-css-grid-container**

```python
data = []
dat=soup.find_all("div",{"class" :"appsumo-css-grid-container"})
for d in dat:
    data.append(d)
data
```

**Tiltes are inside a class : appsumo-deal-title**

```python
titles = []

title = soup.findAll("a",{"class" :"appsumo-deal-title"})
for t in title:
    print(t.text.strip()) 
    titles.append(t.text.strip())

print(titles)
len(titles)
```

**Deal Types are inside span class : deal-plan-type**

```python
deal_types = []

deal = soup.findAll("span",{"class" :"deal-plan-type"})
for d in deal:
    print(d.text.strip()) 
    deal_types.append(d.text.strip())

print(deal_types)
len(deal_types)
```

**Rating are inside span class : appsumo-review-decimal**

```python
ratings = []

rating = soup.findAll("span",{"class" :"appsumo-review-decimal"})
for r in rating:
    print(r.text.strip()) 
    ratings.append(r.text.strip())

print(ratings)
len(ratings)
```

**Number of reviews are inside a class : appsumo-link**

```python
no_of_reviews = []

review = soup.findAll("div",{"class":"appsumo-taco-rating-container"})
for r in review:
    try: 
        rev = r.find("a",{"class" :"appsumo-link"})
        print(rev.text.strip()) 
        no_of_reviews.append(rev.text.replace("Reviews","").strip())
        
    except:
        no_of_reviews.append("N\A")
print(no_of_reviews) 
len(no_of_reviews)
```

**Reviews Links are inside a class : appsumo-link**

```python
review_link = []

review = soup.findAll("div",{"class":"appsumo-taco-rating-container"})
for r in review:
    try:
        rev = r.find("a",{"class" :"appsumo-link"})
        url="https://appsumo.com"+rev['href']
        print(url)
        review_link.append(url)
    except:
        review_link.append("N\A")

print(review_link)
len(review_link)
```

**Deal Story are inside p class : appsumo-deal-story**

```python
dealstory = []

deals = soup.findAll("p",{"class" :"appsumo-deal-story"})
for ds in deals:
    print(ds.text.strip()) 
    dealstory.append(ds.text.strip())

print(dealstory)
len(dealstory)
```

**Old Price are inside span class : appsumo-old-price**

```python
oldprice = []

prices = soup.findAll("span",{"class" :"appsumo-old-price"})
for op in prices:
    print(op.text.strip()) 
    oldprice.append(op.text.strip())

print(oldprice)
len(oldprice)
```

**Discounted Price are inside strong class : appsumo-price**

```python
prices = []

price = soup.findAll("strong",{"class" :"appsumo-price"})
for p in price:
    print(p.text.strip()) 
    prices.append(p.text.strip())

print(prices)
len(prices)
```

**Product Tags are inside div class : appsumo-deal-title-prodcut-tag-container**

```python
product_tags = []

product = soup.findAll("div",{"class":"appsumo-deal-tile-content"})
for p in product:
    tags = p.find("div",{"class" :"appsumo-deal-tile-product-tag-container"})
    print(tags.text.strip()) 
    product_tags.append(tags.text.replace("\n    \n   ","").strip())

print(product_tags)
len(product_tags)
```

**Creating Dictonary**

```python
dict = {
    'Title' : titles,
    'Deal Type' : deal_types,
    'Rating' : ratings,
    'Number of Reviews' : no_of_reviews,
    'Reviews Link':review_link,
    'Deal Story':dealstory,
    'Old Price':oldprice,
    'Discounted Price':prices,
    'Product Tags':product_tags
}
```

**Creating DataFrame**

```python
df = pd.DataFrame.from_dict(dict,orient = 'index').T
df
```

**Exporting To CSV**

```python
#Dataframe to csv
df.to_csv('AppSumo.csv')
```

```python
#Creating Fucntion to clearn the currency

def clean_currency(x):
    """ If the value is a string, then remove currency symbol and delimiters
    otherwise, the value is numeric and can be converted
    """
    if isinstance(x, str):
        return(x.replace('$', '').replace(',', ''))
    return(x)
```
