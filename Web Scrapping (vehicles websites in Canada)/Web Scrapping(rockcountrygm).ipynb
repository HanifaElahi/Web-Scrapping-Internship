```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from selenium import webdriver
import time
from matplotlib import pyplot as plt
```

```python
urla = "https://www.rockcountrygm.ca/VehicleSearchResults?search=new"
```

```python
driver=webdriver.Firefox(executable_path="C:/Users/DELL/Downloads/geckodriver.exe")
#set the value of executable path
driver.get(urla)
```

```python
page_source=driver.page_source
```

```python
#parsing the page
soup=BeautifulSoup(page_source,'html.parser')
```

```python
#All the data present in a page is present in div class : deck
data=soup.findAll('div',{"class":"deck"})
#printing the extracted data of a single page
for info in data:
    print(info.text)
```

```python
make_years = []#taking an empty list
a = soup.findAll('span',{"class":"year"})#scrapping make years from span class : year
for i in a:
    print(i.text)
    make_years.append(i.text)#appending values
print(make_years)#printing make years
len(make_years)
```

```python
#Finding the make of cars using find_all
#All the make are present in span class : make
car_make = []
a=soup.findAll('span',{"class":"make"})
for i in a :
    car_make.append(i.text.replace("\n","").strip())#appending values
print(car_make)#printing make
len(car_make)#printing length of list
```

```python
#Finding the model of cars using find_all
#All the model are present in span class : make
car_model = []
a=soup.findAll('span',{"class":"model"})
for i in a :
    car_model.append(i.text.replace("\n","").strip())#appending values
print(car_model)#printing model
len(car_model)#printing length of list
```

```python
#Finding the condition of cars using find_all
#All the condition are present in span class : condition
condition = []
a=soup.findAll('span',{"class":"condition"})
for i in a :
    condition.append(i.text.replace("\n","").strip())#appending values
print(condition)#printing condition
len(condition)#printing length of list
```

```python
msrp = []
car_price = []
savings = []
# taking empty lists

#All the prices are inside ul class: prices
spec= soup.find_all("ul",{"class": "prices"})

for i in spec:
    newsoup = BeautifulSoup(str(i),'html.parser')#using parser for scrapping inside details
    lis = newsoup.find_all('div')#finding all div inside ul
    for li in lis:
        
        if 'MSRP' in li.text: # if present 
            msrp.append(li.text.replace("\nMSRP\n","").strip()) # then append
        if 'Selling Price' in li.text:# if present
            car_price.append(li.text.replace("\nSelling Price\n\n\n\n","").strip())# then append
        if 'Savings' in li.text:# if present
            savings.append(li.text.replace("\n\n        Savings\n    \n\n        ","").strip())# then append
            
#priniting individual lists
print(msrp)
print(car_price)
print(savings)
```

```python
#creating dict of scrapped data
dict = {
    'Make Year' : make_years , 
    'Make' : car_make,
    'Model':car_model,
    'Condition':condition,
    'MSRP': msrp ,
    'Price' : car_price ,
    'Savings' : savings
}
```

```python
#creating dataframe from dict
df = pd.DataFrame.from_dict(dict,orient = 'index').T
df
```

```python
df['Dealer'] = 'rockcountrygm'
```
