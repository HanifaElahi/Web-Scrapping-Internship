```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from selenium import webdriver
import time
from matplotlib import pyplot as plt
```

```python
urla = "https://www.easternchrysler.com/vehicles/new?st=price,asc&view=grid&sc=new"
```

```python
driver=webdriver.Firefox(executable_path="C:/Users/DELL/Downloads/geckodriver.exe")
#set the value of executable path
driver.get(urla)
```

```python
page_source=driver.page_source
```

```python
#parsing the page
soup=BeautifulSoup(page_source,'html.parser')
```

```python
#All the data present in a page is present in class row
data=soup.findAll('div',{"class":"row"})
#printing the extracted data of a single page
for info in data:
    print(info.text)
```

```python
#Finding the titles of cars using find_all
#All the titles are present in h4 class :vehicle-card__title
car_titles = []
a=soup.findAll('h4',{"class":"vehicle-card__title"})
for i in a :
    print(i.text.strip())#printing titles
    car_titles.append(i.text.strip())#appending values
print(car_titles)#printing titles
len(car_titles)#printing length of list
```

```python
body_style = []
drive_train = []
engine = []
exterior = []
stock = []
transmission = []
mileage = []

# taking empty lists

#All the specifications are inside ul class: detailed-specs inline-view
spec= soup.find_all("ul",{"class": "detailed-specs inline-view"})

for i in spec:
    newsoup = BeautifulSoup(str(i),'html.parser')#using parser for scrapping inside details
    lis = newsoup.find_all('li')#finding all li inside ul
    for li in lis:
        
        if 'Body Style' in li.text: # if present 
            body_style.append(li.text.replace("Body Style: ","")) # then append
        if 'Drive Train' in li.text:# if present
            drive_train.append(li.text.replace("Drive Train: ",""))# then append
        if 'Engine' in li.text:# if present
            engine.append(li.text.replace("Engine: ",""))# then append
        if 'Exterior' in li.text:# if present
            exterior.append(li.text.replace("Exterior Colour: ",""))# then append# then append
        if 'Stock' in li.text:# if present
            stock.append(li.text.replace("Stock #:",""))#then append
        if 'Transmission' in li.text:# if present
            transmission.append(li.text.replace("Transmission:",""))# then append# then append
        if 'Kilometres' in li.text:# if present
            mileage.append(li.text.replace("Kilometres: ",""))#then append
            
#priniting individual lists
print(body_style)
print(drive_train)
print(engine)
print(exterior)
print(stock)
print(transmission)
print(mileage)
```

```python
#Finding the sales price on cars using find_all
#All the  sales price  are present in div class : price-block__single price-block--expanded
car_price = []
c = soup.findAll('div',{"class":"price-block__single price-block--expanded"})
for i in c :
    sp=i.findAll('div',{"class":"price-block__price price-block__price--lg price-block__price--primary"})#digging further
    for j in sp:
        s=j.find('span',{"class":"df aifs"}).text.split("df aifs")[0]#actual values are inside span class :df aifs
        print(s)#printing initial prices
        car_price.append(s)#appending in a list
print(car_price)#printing list
len(car_price)#length of list
```

```python
#Finding the discounted price on cars using find_all
#All the  discounted price  are present in div class : vehicle-card__price-wrap vehicle-card__price-wrap--finance
d_price = []
c = soup.findAll('div',{"class":"vehicle-card__price-wrap vehicle-card__price-wrap--finance"})
for i in c :
    sp=i.findAll('div',{"class":"price-block__single price-block--expanded price-block__single--finance"})#digging further
    for j in sp:
        s=j.find('span',{"class":"df aifs"}).text.split("df aifs")[0]#actual values are inside span class :df aifs
        print(s)#printing initial prices
        d_price.append(s)#appending in a list
print(d_price)#printing list
len(d_price)#length oflist
```

```python
#Finding the msrp of cars using find_all
#All the  msrp are present in div class :price-block__single price-block--expanded price-block--hide-line
msrp = []
c = soup.findAll('div',{"class":"price-block__single price-block--expanded price-block--hide-line"})
for i in c :
    sp=i.findAll('div',{"class":"price-block__price price-block__price--md price-block__price--grey"})#digging further
    for j in sp:
        s=j.find('span',{"class":"df aifs"}).text.split("df aifs")[0]#actual values are inside span class :df aifs
        print(s)#printing msrp
        msrp.append(s)#appending in a list
print(msrp)#printing list
len(msrp)#length oflist

```

```python
#creating dict of scrapped data
dict = {
    'Title' : car_titles ,
    'Body style' : body_style , 
    'Engine': engine ,
    'Exterior' : exterior ,
    'Stock' : stock ,
    'Transmission': transmission,
    'Mileage':mileage,
    'Drive Train': drive_train,
    'MSRP':msrp,
    'Discount':d_price,
    'Price':car_price
}
```

```python
#creating dataframe from dict
df = pd.DataFrame.from_dict(dict,orient = 'index').T
df
```

```python
df['Condition'] = 'New'
df['Dealer'] = 'easternchrysler'
```

```python
split_title = df["Title"].str.split(" ", n = 2, expand = True) 
# splitting Tile column 
```

```python
df["Make Year"]= split_title[0]
```

```python
df["Make"]= split_title[1] 
```

```python
df["Model"]= split_title[2] 
```

```python
df.drop(columns =["Title"], inplace = True) 
```
