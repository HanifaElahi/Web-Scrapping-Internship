```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from selenium import webdriver
import time
from matplotlib import pyplot as plt
```

```python
urla = input(("Enter url : "))
# "https://www.jimgauthierchev.com/VehicleSearchResults?search=new"
```

```python
driver=webdriver.Firefox(executable_path="C:/Users/DELL/Downloads/geckodriver.exe")
#set the value of executable path
driver.get(urla)
```

```python
page_source=driver.page_source
```

```python
#parsing the page
soup=BeautifulSoup(page_source,'html.parser')
```

```python
#All the data present in a page is present in class deck
data=soup.findAll('div',{"class":"deck"})
#printing the extracted data of a single page
for info in data:
    print(info.text)
```

```python
#Finding the titles of cars using find_all
#All the titles are present in h4 template : vehicle-name
car_titles = []
a=soup.findAll('h4',{"template":"vehicle-name"})
for i in a :
    print(i.text)#printing titles
    car_titles.append(i.text.strip())#appending values
print(car_titles)#printing titles
len(car_titles)#printing length of list
```

```python
transmission = []
engine = []
exterior = []
stock = []
drive_train = []
trim= []
# taking empty lists

spec= soup.findAll('ul',{"class":"vehicleIdentitySpecs"}) #all the specifications are inside class  vehicleIdentitySpecs
for i in spec:
    newsoup = BeautifulSoup(str(i),'html.parser')#using parser for scrapping inside details
    lis = newsoup.find_all('li')#finding all li
    for li in lis:
    
        if 'Stock Number' in li.text:
            stock.append(li.text.replace("Stock Number\n","").strip())
        if 'Trim' in li.text:
            trim.append(li.text.replace("Trim\n","").strip())
        if 'Engine' in li.text:
            engine.append(li.text.replace("Engine\n\n\n","").strip())
        if 'Drive Wheels' in li.text:
            drive_train.append(li.text.replace("Drive Wheels\n","").strip())
        if 'Transmission' in li.text:
            transmission.append(li.text.replace("Transmission\n\n","").strip())
            
print(stock)
print(trim)
print(engine)
print(drive_train)
print(transmission)
```

```python
msrp = []
car_price = []
savings = []
# taking empty lists

spec= soup.findAll('ul',{"not":"enableOptionalSpecialsMathBox"})  #main class
for i in spec:
    newsoup = BeautifulSoup(str(i),'html.parser')#using parser for scrapping inside details
    lis = newsoup.find_all('li')#finding all li
    for li in lis:
        
        if 'Priced To Sell For' in li.text:
            msrp.append(li.text.replace("\n\n\n\n\n    Priced To Sell For\n\n\n    ","").strip())
         
        if 'Final Sale Price' in li.text:
            car_price.append(li.text.replace("\nFinal Sale Price:\n","").strip())
        
        if 'Total Savings' in li.text:
            savings.append(li.text.replace("\nTotal Savings\n","").strip())

print(msrp)
print(car_price)
print(savings)
```

```python
#creating dict of scrapped data
dict = {
    'Title' : car_titles,
    'Transmission' : transmission,
    'Trim':trim,
    'Stock':stock,
    'Engine' : engine ,
    'Drive Train':drive_train,
    'Transmission':transmission,
    'MSRP':msrp,
    'Price' : car_price,
    'Savings':savings
}
```

```python
df = pd.DataFrame.from_dict(dict,orient = 'index').T
df
```

```python
#Dataframe to csv
df.to_csv('jimgautheirchervolet.csv')
```
