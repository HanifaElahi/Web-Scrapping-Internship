```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from selenium import webdriver
import time
from matplotlib import pyplot as plt
```

```python
urla = "https://autocatch.com/used-cars-classifieds?city=Winnipeg&province=Manitoba"
```

```python
driver=webdriver.Firefox(executable_path="C:/Users/DELL/Downloads/geckodriver.exe")
#set the value of executable path
driver.get(urla)
```

```python
page_source=driver.page_source
```

```python
#parsing the page
soup=BeautifulSoup(page_source,'html.parser')
```

```python
#All the data present in a page is present in div class :  columns large-8
data=soup.findAll('div',{"class":"columns large-8"})
#printing the extracted data of a single page
for info in data:
    print(info.text)
```

```python
#Finding the titles of cars using find_all
#All the titles are present in h6 class : vehicle-name
car_titles = []
a=soup.findAll('h6',{"class":"vehicle-name"})
for i in a :
    print(i.text)#printing titles
    car_titles.append(i.text.strip())#appending values
print(car_titles)#printing titles
len(car_titles)#printing length of list
```

```python
#Finding the status of cars using find_all
#All the status  are present in div class : vehicle-status
car_status = []
a=soup.findAll('div',{"class":"vehicle-status"})
for i in a :
    print(i.text)#printing status 
    car_status.append(i.text.replace("Vehicle Status","").strip())#appending values
print(car_status)#printing status
len(car_status)#printing length of list
```

```python
engine = []
stock = []
transmission = []
# taking empty lists

spec= soup.findAll('div',{"class":"detail-list hide-for-small-only"})#all the specs are inside div class : detail-list hide-for-small-only

for i in spec:
    newsoup = BeautifulSoup(str(i),'html.parser')#using parser for scrapping inside details
    lis = newsoup.find_all('li')#finding all li inside div
    for li in lis:
        
        if 'Engine' in li.text: #if present
            engine.append(li.text.replace("Engine: ",""))#then append
        if 'Stock' in li.text:#if present
            stock.append(li.text.replace("Stock#:",""))#then append
        if 'Transmission' in li.text:#if present
            transmission.append(li.text.replace("Transmission: ",""))#then append
        
#printing individual lists        
print(engine)
print(stock)
print(transmission)
```

```python
#Finding the car price on cars using find_all
#All the  car price  are present in div class : price
car_price = []
mileage = []
c = soup.findAll('div',{"class":"price"})
for i in c :
    if 'Price' in i.text: #if present
        car_price.append(i.text.replace("Price\n","").strip()) #then append
    if 'Mileage' in i.text:#if present
        mileage.append(i.text.replace("Mileage  ","").strip()) #then append

 # printing individual lists
print(car_price)
print(mileage)
```

```python
#creating dict of scrapped data
dict = {
    'Title' : car_titles,
    'Price' : car_price,
    'Mileage':mileage,
    'Engine' : engine ,
    'Stock' : stock,
    'Transmission' : transmission,
    'Condition' : car_status
}
```

```python
df = pd.DataFrame.from_dict(dict,orient = 'index').T
df
```

```python
split_title = df["Title"].str.split(" ", n = 3, expand = True) 
```

```python
df["Make Year"]= split_title[0] 
```

```python
df["Make"]= split_title[1] 
```

```python
df['Model']= split_title[2]
```

```python
df.drop(columns =["Title"], inplace = True) 
```
