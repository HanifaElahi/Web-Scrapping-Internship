```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from selenium import webdriver
import time
from matplotlib import pyplot as plt
```

```python
urla = "https://www.portagetoyota.com/inventory/search?page=1&page_length=20&make=Toyota&stock_type=New"
```

```python
driver=webdriver.Firefox(executable_path="C:/Users/DELL/Downloads/geckodriver.exe")
#set the value of executable path
driver.get(urla)
```

```python
page_source=driver.page_source
```

```python
#parsing the page
soup=BeautifulSoup(page_source,'html.parser')
```

```python
#All the data present in a page is present in class group-content
data=soup.findAll('section',{"class":"group-content"})
#printing the extracted data of a single page
for info in data:
    print(info.text)
```

```python
#Finding the titles of cars using find_all
#All the titles are present in h4 itemprop : name
car_titles = []
a=soup.findAll('h4',{"itemprop":"name"})
for i in a :
    print(i.text)
    car_titles.append(i.text)#appending values
print(car_titles)#printing titles
len(car_titles)#printing length of list
```

```python
#Finding the price of cars using find_all
#All the price are present in strong class : price _bpcolor
car_price = []
a=soup.findAll('strong',{"class":"price _bpcolor"})
for i in a :
    print(i.text)
    car_price.append(i.text.strip())#appending values
print(car_price)#printing price
len(car_price)#printing length of list
```

```python
mileage = [] #taking an empty list
a = soup.findAll('dd',{"class":"srp-odometer"})#scrapping
for i in a:
    print(i.text)
    mileage.append(i.text)#appending values
print(mileage)#printing length of list
```

```python
exterior = []#taking an empty list
a = soup.findAll('dd',{"class":"srp-color-exterior"})#scrapping
for i in a:
    print(i.text)
    exterior.append(i.text)#appending values
print(exterior)#printing length of list
```

```python
engine = []#taking an empty list
a = soup.findAll('dd',{"class":"srp-engine-combined"})#scrapping
for i in a:
    print(i.text)
    engine.append(i.text)#appending values
print(engine)#printing length of list
```

```python
drive_train = []#taking an empty list
a = soup.findAll('dd',{"class":"srp-driveType"})
for i in a:
    print(i.text)
    drive_train.append(i.text)#appending values
print(drive_train)#printing list
```

```python
stock = []#taking an empty list
a = soup.findAll('dd',{"class":"srp-stock-number"})#scrapping
for i in a:
    print(i.text)
    stock.append(i.text)#appending values
print(stock)#printing list
```

```python
vin = []#taking an empty list
a = soup.findAll('dd',{"class":"srp-vin"})#scrapping
for i in a:
    print(i.text)
    vin.append(i.text)#appending values
print(vin)#printing list
```

```python
#Creating a dict of scrapped data
dict = {
        'Title' : car_titles , 
        'Mileage':mileage,
        'Exterior':exterior,
        'Engine':engine,
        'Drive Train':drive_train,
        'Stock':stock,
        'VIN':vin,
        'Price':car_price
}
```

```python
df = pd.DataFrame.from_dict(dict,orient = 'index').T
df
```

```python
df['Dealer'] = 'portagetoyota'
df['Condition'] =  'New'
```

```python
split_title = df["Title"].str.split(" ", n = 2, expand = True) 
# splitting Tile column 
df["Make Year"]= split_title[0] 
df["Make"]= split_title[1] 
df["Model"]= split_title[2] 
df
```

```python
df.drop(columns =["Title"], inplace = True)
df
```

```python
#Dataframe to csv
df.to_csv('portagetoyota.csv')
```
