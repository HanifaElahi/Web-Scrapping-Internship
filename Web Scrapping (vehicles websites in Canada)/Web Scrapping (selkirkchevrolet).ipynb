```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from selenium import webdriver
import time
from matplotlib import pyplot as plt
```

```python
urla = input(("Enter url : "))
# "https://www.selkirkchevrolet.com/new-vehicles/"
```

```python
driver=webdriver.Firefox(executable_path="C:/Users/DELL/Downloads/geckodriver.exe")
#set the value of executable path
driver.get(urla)
```

```python
page_source=driver.page_source
```

```python
#parsing the page
soup=BeautifulSoup(page_source,'html.parser')
```

```python
#All the data present in a page is present in id lvrp-results-container
data=soup.findAll('div',{"id":"lvrp-results-container"})
#printing the extracted data of a single page
for info in data:
    print(info.text)
```

```python
#Finding the titles of cars using find_all
#All the titles are present in h2 class : result-title
car_titles = []
a=soup.findAll('h2',{"class":"result-title"})
for i in a :
    print(i.text)#printing titles
    car_titles.append(i.text)#appending values
print(car_titles)#printing titles
len(car_titles)#printing length of list
```

```python
exterior = []
interior = []
# taking empty lists

spec= soup.findAll('ul',{"class":"vehicle-details"}) #scrappig details inside class:vehicle-details 

for i in spec:
    newsoup = BeautifulSoup(str(i),'html.parser')#using parser for scrapping inside details
    lis = newsoup.find_all('li')#finding all li inside ul
    for li in lis:
        
        if 'Exterior' in li.text: #if present
            exterior.append(li.text.replace("Exterior:",""))#then append
        if 'Interior' in li.text:#if present
            interior.append(li.text.replace("Interior:",""))#then append
       
        
#printing individual lists        
print(exterior)
print(interior)
```

```python
msrp = [] #taking empty list
a = soup.findAll('div',{"class":"price-block origional"})#scrapping
for i in a:
    print(i.text.strip())
    msrp.append(i.text.replace("MSRP\n","").strip())#appending in a list
print(msrp)#printing list
```

```python
rebate = []#taking empty list
a = soup.findAll('div',{"class":"price-block"})#scrapping
for i in a:
    if 'Rebate' in i.text:
        rebate.append(i.text.replace("\n\nRebate\n","").strip())#appending in a lis
print(rebate)#printing list
```

```python
discount = []#taking empty list
a = soup.findAll('div',{"class":"price-block"})#scrapping
for i in a:
    
    if 'Discount' in i.text:
        discount.append(i.text.replace("Rural Discount\n","").strip())#appending in a list
    
print(discount)#printing list
```

```python
car_price = []#taking empty list
a = soup.findAll('div',{"class":"price-block our-price real-price"})#scrapping
for i in a:
    print(i.text.strip())
    car_price.append(i.text.replace("Price","").strip())#appending in a list
print(car_price) #printing list
```

```python
#creating dict of scrapped data
dict = {
    'Title' : car_titles ,
    'Exterior' : exterior , 
    'Interior ': interior,
    'MSRP' : msrp ,
    'Rebate' : rebate ,
    'Discount': discount,
    'Price' : car_price
}
```

```python
df = pd.DataFrame.from_dict(dict,orient = 'index').T
df
```

```python
#Dataframe to csv
df.to_csv('selkirkchevrolet.csv')
```
