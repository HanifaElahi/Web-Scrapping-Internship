```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from selenium import webdriver
import time
from matplotlib import pyplot as plt
```

```python
urla = input(("Enter url : "))
# "https://www.marshallusedcars.ca/used-cars-brandon?ps=48&SortBy=Make&SortOrder=ASC"
```

```python
driver=webdriver.Firefox(executable_path="C:/Users/DELL/Downloads/geckodriver.exe")
#set the value of executable path
driver.get(urla)
```

```python
page_source=driver.page_source
```

```python
#parsing the page
soup=BeautifulSoup(page_source,'html.parser')
```

```python
#All the data present in a page is present in class col-sm-9 mb20
data=soup.findAll('div',{"class":"col-sm-9 mb20"})
#printing the extracted data of a single page
for info in data:
    print(info.text.replace("\n",""))
```

```python
#Finding the titles of cars using find_all
#All the titles are present in h2 clss : eziVehicleName
car_titles = []
a=soup.findAll('h2',{"class":"eziVehicleName"})
for i in a :
    print(i.text)#printing titles
    car_titles.append(i.text)#appending values
print(car_titles)#printing titles
len(car_titles)#printing length of list
```

```python
#Finding the price of cars using find_all
#All the price are present in span class: eziPriceValue
car_price = []
a=soup.findAll('span',{"class":"eziPriceValue"})
for i in a :
    print(i.text)#printing car price
    car_price.append(i.text)#appending values
print(car_price)#printing car price
len(car_price)#printing length of list
```

```python
mileage = []
stock = []
body_type = []
transmission = []
engine = []
cylinders = []
fuel_type = []
drive_type = []
doors = []
seats = []
exterior = []
interior = []
#taking empty individual lists

spec = soup.find_all('div',{"class":"col-md-6"}) #all the details are inside div class: col-md-6
for i in spec:
    
    if 'Odometer' in i.text:#if present
        mileage.append(i.text.replace("Odometer:",""))#then append
    if 'Stock' in i.text:#if present
        stock.append(i.text.replace("Stock #:",""))#then append
    if 'Body' in i.text:#if present
        body_type.append(i.text.replace("Body:",""))#then append
    if 'Trans' in i.text:#if present
        transmission.append(i.text.replace("Trans:",""))#then append
    if 'Engine' in i.text:#if present
        engine.append(i.text.replace("Engine:",""))#then append
    if 'Fuel' in i.text:#if present
        fuel_type.append(i.text.replace("Fuel:",""))#then append
    if 'Drive' in i.text:#if present
        drive_type.append(i.text.replace("Drive:",""))#then append
    if 'Doors' in i.text:#if present
        doors.append(i.text.replace("Doors:",""))#then append
    if 'Seats' in i.text:#if present
        seats.append(i.text.replace("Seats:",""))#then append
    if 'Exterior' in i.text:#if present
        exterior.append(i.text.replace("Exterior:",""))#then append
    if 'Interior' in i.text:#if present
        interior.append(i.text.replace("Interior:",""))#then append
    if 'Cylinders' in i.text:#if present
        cylinders.append(i.text.replace("Cylinders:",""))#then append
    

#printing individual lists
print(mileage) 
print(stock)
print(body_type)
print(transmission)
print(engine)
print(fuel_type)
print(drive_type)
print(doors) 
print(seats)
print(exterior)
print(interior)
print(cylinders)
```

```python
#creating dict of scrapped data
dict = {
    'Title' : car_titles,
    'Price' : car_price,
    'Mileage': mileage,
    'Stock' : stock,
    'Body Type':body_type,
    'Transmission': transmission,
    'Engine':engine,
    'Fuel Type':fuel_type,
    'Drive Type':drive_type,
    'Doors':doors,
    'Seats':seats,
    'Exterior':exterior,
    'Interior':interior,
    'Cylinders':cylinders
}
```

```python
df = pd.DataFrame.from_dict(dict,orient = 'index').T
df
```

```python
#Dataframe to csv
df.to_csv('marshallusedcars.csv')
```
