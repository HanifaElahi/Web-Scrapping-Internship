```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from selenium import webdriver
import time
from matplotlib import pyplot as plt
```

```python
urla = "https://www.wilfselieford.ca/used-vehicles-just-west-of-winnipeg"
```

```python
driver=webdriver.Firefox(executable_path="C:/Users/DELL/Downloads/geckodriver.exe")
#set the value of executable path
driver.get(urla)
```

```python
page_source=driver.page_source
```

```python
#parsing the page
soup=BeautifulSoup(page_source,'html.parser')
```

```python
#All the data present in a page is present in class vehicle-list center-block max-1200
data=soup.findAll('div',{"class":"vehicle-list center-block max-1200"})
#printing the extracted data of a single page
for info in data:
    print(info.text)
```

```python
#Finding the titles of cars using find_all
#All the titles are present in h4 class : vehicle-header media-heading
car_titles = []
a=soup.findAll('h4',{"class":"vehicle-header media-heading"})
for i in a :
    print(i.text)#printing titles
    car_titles.append(i.text)#appending values
print(car_titles)#printing titles
len(car_titles)#printing length of list
```

```python
#Finding the description of cars using find_all
#All the description are present in div class : vehicle-highlights-tagline
car_description = []
b = soup.findAll('div',{"class":"vehicle-highlights-tagline"})
for i in b :
    print(i.text)
    car_description.append(i.text)#apppending inside a separate list
print(car_description)#printing list
len(car_description)#length of list
```

```python
#Finding the sales price on cars using find_all
#All the  sales price  are present in div class : vehicle-price-full
car_price = []
c = soup.findAll('div',{"class":"vehicle-price-full"})
for i in c :
    print(i.text)
    car_price.append(i.text.replace("* Price",""))#appending in a list
print(car_price)#printing list
len(car_price)#length of list


```

```python

certification = []#taking empty list

c = soup.findAll('div',{"class":"vehicle-specs"})#scrapping cerification
for i in c :
    a = i.find('span').text.split('|')[0]#finding using split
    certification.append(a)#appending in lisy
    
print(certification)#printing list

 
```

```python
#creating dict of scrapped data
dict = {
    'Title' : car_titles ,
    'Description' : car_description , 
    'Certification ': certification ,
    'Price' : car_price
}
```

```python
df = pd.DataFrame.from_dict(dict,orient = 'index').T
df
```

```python
df['Dealer'] = 'wilfselieford'
df['Condition'] = 'Used'
```

```python
split_title = df["Title"].str.split(" ", n = 2, expand = True) 
# splitting Tile column 
df["Make Year"]= split_title[0] 
df["Make"]= split_title[1] 
df["Model"]= split_title[2] 
df
```

```python
df.drop(columns =["Title"], inplace = True)
df
```

```python
#Dataframe to csv
df.to_csv('welfselieford.csv')
```
