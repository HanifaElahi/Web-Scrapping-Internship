```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from selenium import webdriver
import time
from matplotlib import pyplot as plt
```

```python
urla = "https://www.swanriverkeychev.com/VehicleSearchResults?search=preowned"
```

```python
driver=webdriver.Firefox(executable_path="C:/Users/DELL/Downloads/geckodriver.exe")
#set the value of executable path
driver.get(urla)
```

```python
page_source=driver.page_source
```

```python
#parsing the page
soup=BeautifulSoup(page_source,'html.parser')
```

```python
#All the data present in a page is present in class deck
data=soup.findAll('div',{"class":"deck"})
#printing the extracted data of a single page
for info in data:
    print(info.text)
```

```python
car_titles = []#taking an empty list
a = soup.findAll('h4',{"template":"vehicle-name"})#titles are inside h4 template : vehicle-name
for i in a:
    print(i.text)
    car_titles.append(i.text.strip())#appending in a list
print(car_titles)#printing list
```

```python
car_price = []#taking an empty list
a = soup.findAll('span',{"class":"value big-and-bold"}) #prices are inside span class : value big-and-bold
for i in a:
    print(i.text)
    car_price.append(i.text)#appending in a list
print(car_price)#printing list
```

```python
transmission = []
engine = []
exterior = []
stock = []
drive_train = []
trim= []
# taking empty lists

spec= soup.findAll('ul',{"class":"vehicleIdentitySpecs"}) #all the specifications are inside class  vehicleIdentitySpecs
for i in spec:
    newsoup = BeautifulSoup(str(i),'html.parser')#using parser for scrapping inside details
    lis = newsoup.find_all('li')#finding all li
    for li in lis:
    
        if 'Stock Number' in li.text: #if present
            stock.append(li.text.replace("Stock Number\n","").strip())#then append
        if 'Trim' in li.text:#if present
            trim.append(li.text.replace("Trim\n","").strip())#then append
        if 'Engine' in li.text:#if present
            engine.append(li.text.replace("Engine\n\n\n","").strip())#then append
        if 'Drive Wheels' in li.text:#if present
            drive_train.append(li.text.replace("Drive Wheels\n","").strip())#then append
        if 'Transmission' in li.text:#if present
            transmission.append(li.text.replace("Transmission\n\n","").strip())#then append

#printing individual lists
print(stock)
print(trim)
print(engine)
print(drive_train)
print(transmission)
```

```python
#creating dict of scrapped data
dict = {
    'Title' : car_titles ,
    'Transmission':transmission, 
    'Price': car_price ,
    'Engine' : engine,
    'Exterior': exterior,
    'Stock' : stock ,
    'Drive Train':drive_train,
    'Trim':trim
}
```

```python
#creating dataframe from dict
df = pd.DataFrame.from_dict(dict,orient = 'index').T
df
```

```python
df['Dealer'] = 'swanriverkeychev'
```

```python
split_title = df["Title"].str.split("\n", n = 8, expand = True) 
# splitting Tile column 
```

```python
df["Condition"]= split_title[0]
```

```python
df["Make Year"]= split_title[1]
```

```python
df["Make"]= split_title[4]
```

```python
df["Model"]= split_title[5]
```
