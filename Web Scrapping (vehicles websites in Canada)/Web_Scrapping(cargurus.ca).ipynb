```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from selenium import webdriver
from matplotlib import pyplot as plt
```

```python
urla = "https://www.cargurus.ca/Cars/inventorylisting/viewDetailsFilterViewInventoryListing.action?sourceContext=carGurusHomePageModel&entitySelectingHelper.selectedEntity=m37&zip=L3Y"
```

```python
driver = webdriver.Firefox(executable_path="C:/Users/DELL/Downloads/geckodriver.exe")
#set the value of executable path
driver.get(urla)
```

```python
page_source=driver.page_source
```

```python
#parsing the page
soup=BeautifulSoup(page_source,'html.parser')
soup
```

```python
#All the data present in a page is present in div class : _3LnDeD
data=soup.findAll('div',{"class":"_3LnDeD"})
#printing the extracted data of a single page
for info in data:
    print(info.text)
```

```python
#Finding the titles of cars using find_all
#All the titles are present in a class : _4Pn4Gb _2yn1Hf
car_titles = [] #taking an empty list
a=soup.findAll('a',{"class":"_4Pn4Gb _2yn1Hf"})
for i in a :
    titles=i.find('h4').text.split("listing")[0] #finding titles inside the h4 heading of class _4Pn4Gb _2yn1Hf
    print(titles) #printing titles
    car_titles.append(titles)#appending in an empty list
print(car_titles)#printing list
len(car_titles)#length of list
```

```python
#Finding the prices of cars using find_all
#All the prices are present in span class : _8E6guA
price = soup.find_all("span",{"class" :"_8E6guA"})
```

```python
#taking an empty list
car_price = []
for bp in price:
    print(bp.text)# printing values
    #appending values in a separate empty list
    car_price.append(bp.text)  
print(car_price)#printing car prices list
len(car_price) #length of list
```

```python
#Finding the mileage of cars using find_all
#All the mileage are present in p class : eMnkbf
mileage=soup.find_all("p",{"class" :"eMnkbf"})
```

```python
#taking an empty list
car_mileage = []
for bm in mileage:
    print(bm.text)# printing values
    car_mileage.append(bm.text) #appending values in an empty list 
print(car_mileage)#printing list
len(car_mileage)#printing length of list
    
```

```python
#Finding the type of deal for cars using find_all
#All the type of deal for cars are present in span class : _2MGfxG _4XnLpU
t_deal = soup.find_all("span",{"class" :"_2MGfxG _4XnLpU"})
```

```python
#taking an empty list
deal_type = []
for d in t_deal:
    print(d.text) #printing values
    deal_type.append(d.text)#appending values
print(deal_type)#printing list
len(deal_type)#length of list
```

```python
#Finding the deal for cars using find_all
#All the  deal for cars are present in span class : _5H2mos
deal = soup.find_all("span",{"class" :"_5H2mos"})
```

```python
#taking an empty list
deal_price = []
for d in deal:
    print(d.text) #printing values
    deal_price.append(d.text) #appending in an empty list
print(deal_price) #printing list
len(deal_price)#length of list
```

```python
#Finding the rating for cars using find_all
#All the  rating for cars are present in span class : _4bosXj
rating = soup.find_all("span",{"class" :"_4bosXj"})
```

```python
#taking an empty list
car_rating = []
for r in rating:
    print(r.text)#printing values
    car_rating.append(r.text)#appending in an empty list
print(car_rating)#printing list
len(car_rating)#length of list
```

```python
#Finding the location of cars using find_all
#All the  location of cars are present in div class : _5zLBkD
location = soup.find_all("div",{"class" :"_5zLBkD"})
```

```python
#Taking an empty list
car_location=[]
for l in location:
    print(l.text)#printing values
    car_location.append(l.text)#appending in an empty list
print(car_location)#printing list
len(car_location)#length of list
```

```python
#Taking an empty list
contacts = []
#Finding the contacts using find_all
#All the  contacts are present in div class : _2jBiNz _2tyPai
c = soup.findAll("div",{"class" :"_2jBiNz _2tyPai"})
for i in c:
    contact = i.find('div').text # contact no is in div tag 
    print(contact)
    contacts.append(contact)#appending in an empty list
print(contacts)#printing contacts
len(contacts)#length of contacts
    
```

```python
#Creating dictionary of the scrapped details
dict = {
    'Title' : car_titles,
    'Contact No.' : contacts, 
    'Price' : car_price,
    'Mileage' : car_mileage,
    'Location': car_location,
    'Deal Type' : deal_type,
    'Deal' : deal_price,
    'Rating' : car_rating
      
}
```

```python
#creating dataframe from dict
df = pd.DataFrame.from_dict(dict,orient = 'index').T
```

```python
df['Condition'] = 'Used'
df['Dealer'] = 'cargurus.ca'
```

```python
split_title = df["Title"].str.split(" ", n = 2, expand = True) 
# splitting Tile column 
```

```python
df['Make Year'] = split_title[0]
```
