```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from selenium import webdriver
import time
from matplotlib import pyplot as plt
```

```python
urla = "https://www.autolistofcanada.com/vehicles/?fn=&pr=&mk=&md=&tr=&bs=&od=&yr=&v1=Passenger%20Vehicles&st=make,desc&view=grid"
```

```python
driver=webdriver.Firefox(executable_path="C:/Users/DELL/Downloads/geckodriver.exe")
#set the value of executable path
driver.get(urla)
```

```python
page_source=driver.page_source
```

```python
#parsing the page
soup=BeautifulSoup(page_source,'html.parser')
```

```python
#All the data present in a page is present in div class : row grid row
data = soup.findAll('div',{"class":"row grid-row"})
#printing the extracted data of a single page
for info in data:
    print(info.text)
```

```python
#Finding the titles of cars using find_all
#All the titles are present in div class : vehicle-card__full-details
car_titles = []
a=soup.findAll('div',{"class":"vehicle-card__full-details"})
for i in a :
    titles=i.find('h3')
    print(titles.text)#printing titles
    car_titles.append(titles.text)
print(car_titles)#printing titles
len(car_titles)#printing length of list
```

```python
body_style = []
mileage = []
transmission = []
drive_train = []
exterior = []
# taking empty lists

#All the specifications are inside ul class: detailed-specs 
spec= soup.find_all("ul",{"class": "detailed-specs"})

for i in spec:
    newsoup = BeautifulSoup(str(i),'html.parser')#using parser for scrapping inside details
    lis = newsoup.find_all('li')#finding all li inside ul
    for li in lis:
        if 'Body Style' in li.text: # if present 
            body_style.append(li.text.replace("Body Style:","")) # then append
        if 'Kilometres' in li.text:# if present
            mileage.append(li.text.replace("Kilometres:",""))# then append
        if 'Transmission' in li.text:# if present
            transmission.append(li.text.replace("Transmission:",""))# then append# then append
        if 'Drive Train' in li.text:# if present
            drive_train.append(li.text.replace("Drive Train:",""))# then append
        if 'Exterior Colour' in li.text:# if present
            exterior.append(li.text.replace("Exterior Colour:",""))#then append
            
#priniting individual lists
print(body_style)
print(mileage)
print(transmission)
print(drive_train)
print(exterior)
```

```python
#Finding the msrp of cars using find_all
#All the msrp are present in div class : vehicle-card__price-wrap
msrp = []
c = soup.findAll('div',{"class":"vehicle-card__price-wrap"})
for i in c :
    sp=i.findAll('div',{"class":"price-block__price price-block__price--md price-block__price--grey"})#digging further
    for j in sp:
        s=j.find('span',{"class":"df aifs"}).text.split("df aifs")[0]#actual values are inside span class :df aifs
        print(s)#printing msrp
        msrp.append(s)#appending in a list
print(msrp)#printing list
len(msrp)#length oflist
```

```python
#Finding the savings using find_all
#All the savings are present in div class : vehicle-card__price-wrap
savings = []
c = soup.findAll('div',{"class":"vehicle-card__price-wrap"})
for i in c :
    sp=i.findAll('div',{"class":"price-block__price price-block__price--sm price-block__price--primary"})#digging further
    for j in sp:
        s=j.find('span',{"class":"df aifs"}).text.split("df aifs")[0]#actual values are inside span class :df aifs
        print(s)#printing savings
        savings.append(s)#appending in a list
print(savings)#printing list
len(savings)#length oflist
```

```python
#Finding the prices of cars using find_all
#All the prices are present in div class : vehicle-card__price-wrap
car_price = []
c = soup.findAll('div',{"class":"vehicle-card__price-wrap"})
for i in c :
    sp=i.findAll('div',{"class":"price-block__price price-block__price--lg price-block__price--dark"})#digging further
    for j in sp:
        s=j.find('span',{"class":"df aifs"}).text.split("df aifs")[0]#actual values are inside span class :df aifs
        print(s)#printing prices
        car_price.append(s)#appending in a list
print(car_price)#printing list
len(car_price)#length oflist
```

```python
#Finding the finance prices of cars using find_all
#All the finance prices are present in div class : vehicle-card__price-wrap
car_f_price = []
c = soup.findAll('div',{"class":"vehicle-card__price-wrap"})
for i in c :
    sp=i.findAll('div',{"class":"price-block__single price-block--expanded price-block__single--finance"})#digging further
    for j in sp:
        s=j.find('span',{"class":"df aifs"}).text.split("df aifs")[0]#actual values are inside span class :df aifs
        print(s)#printing prices
        car_f_price.append(s)#appending in a list
print(car_f_price)#printing list
len(car_f_price)#length oflist
```

```python
#Finding the duration  using find_all
#All the duration are present in div class : vehicle-card__price-wrap
duration = []
c = soup.findAll('div',{"class":"vehicle-card__price-wrap"})
for i in c :
    sp=i.findAll('div',{"class":"price-block__single price-block--expanded price-block__single--finance"})#digging further
    for j in sp:
        s=j.find('span',{"class":"price-block__sub"}).text.replace("\t","")#actual values are inside span class :df aifs
        print(s)#printing duration
        duration.append(s.strip())#appending in a list
print(duration)#printing list
len(duration)#length oflist
```

```python
#creating dict of scrapped data
dict = {
    'Title' : car_titles ,
    'Mileage' : mileage , 
    'Body Style' :body_style,
    'Exterior' : exterior ,
    'Drive Train' : drive_train ,
    'Transmission' : transmission,
    'MSRP':msrp,
    'Price': car_price ,
    'Finance Price per week': car_f_price,
    'Finance Duration':duration
}
```

```python
#creating dataframe from dict
df = pd.DataFrame.from_dict(dict,orient = 'index').T
df
```

```python
split_title = df["Title"].str.split(" ", n = 3, expand = True) 
# splitting Tile column 
```

```python
df["Make Year"]= split_title[0] 
```

```python
df["Make"]= split_title[1] 
```

```python
df["Model"]= split_title[2] 
```

```python
df.drop(columns =["Title"], inplace = True) 
```
