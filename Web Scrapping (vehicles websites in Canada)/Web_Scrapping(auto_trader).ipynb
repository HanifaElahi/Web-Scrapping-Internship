```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from selenium import webdriver
import time
from matplotlib import pyplot as plt
```

```python
urla = "https://www.autotrader.ca/cars/qc/montr%C3%A9al-nord/?rcp=0&rcs=0&prx=100&prv=Quebec&loc=H1G%205P2&hprc=True&wcp=True&sts=New-Used&inMarket=basicSearch&make=Audi"
```

```python
driver=webdriver.Firefox(executable_path="C:/Users/DELL/Downloads/geckodriver.exe")
#set the value of executable path
driver.get(urla)
```

```python
page_source=driver.page_source
```

```python
#parsing the page
soup=BeautifulSoup(page_source,'html.parser')
```

```python
#All the data present in a page is present in  div class : col-xs-12 result-item-inner
data=soup.findAll('div',{"class":"col-xs-12 result-item-inner"})
#printing the extracted data of a single page
for info in data:
    print(info.text)
```

```python
#Finding the titles of cars using find_all
#All the titles are present in div class : result-title click
titles=soup.find_all("a",{"class" :"result-title click"})
```

```python
#taking an empty list
car_titles = []
#printing all the separate titles using for loop
for title in titles:
    print(title.text)# printing values
    #appending values in a separate empty list
    #there are several \n \n in extracted titles therefore stripping them using .strip()
    car_titles.append(title.text.strip())  
print(car_titles)#printing car titles list
len(car_titles)
```

```python
#Finding the price of cars using find_all
#All the prices are present in div class : price-amount
price_amount=soup.find_all("div",{"class" :"price"})
```

```python
#taking an empty list
car_price=[]
#printing all the separate prices using for loop
for p in price_amount:
    s = p.find('span').split('')[0]
    print(s.text)# printing values
    #appending values in a separate empty list
    #there are several \n \n in extracted prices therefore stripping them using .strip()
    car_price.append(s.text)    
print(car_price)#printing car prices list
len(car_price)
```

```python
#Finding the details of cars using find_all
#All the details  are present in div class : details used
details=soup.find_all("p",{"class" :"details used"})
```

```python
#taking an empty list
car_details = []
#printing all the separate details using for loop
for detail in details:
    print(detail.text)# printing values
    #appending values in a separate empty list
    #there are several \n \n in extracted details therefore stripping them using .strip()
    car_details.append(detail.text.strip())
print(car_details)#printing car details list
len(car_details)
```

```python
#Finding the proximity of cars using find_all
#All the proximity values are present in div class : proximity-text
proximity_text=soup.find_all("span",{"class" :"proximity-text"})
```

```python
#taking an empty list
car_prox = []
#printing all the separate proximity values using for loop
for proximity in proximity_text:
    print(proximity.text) # printing values
    #appending values in a separate empty list
    #there are several \n \n in extracted proximity therefore stripping them using .strip()
    car_prox.append(proximity.text)   
print(car_prox)#printing car proximity list
len(car_prox)
```

```python
#Finding the mileage of cars using find_all
#All the mileage values are present in div class : kms
mileage=soup.find_all("div",{"class" :"kms"})
```

```python
#taking an empty list
car_mileage=[]
#printing all the separate mileage values using for loop
for m in mileage:
    print(m.text) # printing values
    #appending values in a separate empty list
    #there are several \n \n in extracted mileages therefore stripping them using .strip()
    car_mileage.append(m.text.replace("Mileage","").strip())    
print(car_mileage) #printing car mileage list
len(car_mileage)
```

```python
#Creating dictionary of scrapped data(Car Title, Car Details, Car Price, Car mileage, Car proximity)
#All these values are in individual lists
dict = {'Title' : car_titles , 'Details' : car_details , 'Price' : car_price , 'Mileage' : car_mileage, 'Proximity ': car_prox}
```

```python
#Creating dataframe from dict
df = pd.DataFrame.from_dict(dict,orient = 'index').T
```

```python
#Displaying dataframe
df
```

```python
df['Condition'] = ''
df['Dealer'] = 'autotrader.ca'
```

```python
split_title = df["Title"].str.split(" ", n = 2, expand = True) 
# splitting Tile column
```

```python
df["Make Year"]= split_title[0] 
```

```python
df["Make"]= split_title[1] 
```

```python
df["Model"]= split_title[2] 
```

```python
df.drop(columns =["Title"], inplace = True) 
```
