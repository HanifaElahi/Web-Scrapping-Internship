```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from selenium import webdriver
import time
from matplotlib import pyplot as plt
```

```python
urla = "https://wpsteinbachautodealers.strathcomcms.com/inventory/search"
```

```python
driver=webdriver.Firefox(executable_path="C:/Users/DELL/Downloads/geckodriver.exe")
#set the value of executable path
driver.get(urla)
```

```python
page_source=driver.page_source
```

```python
#parsing the page
soup=BeautifulSoup(page_source,'html.parser')
```

```python
#All the data present in a page is present in class group-content
data=soup.findAll('section',{"class":"group-content"})
#printing the extracted data of a single page
for info in data:
    print(info.text)
```

```python
#Finding the titles of cars using find_all
#All the titles are present in h4 itemprop : name
car_titles = []
a=soup.findAll('h4',{"itemprop":"name"})
for titles in a :
    print(titles.text)#printing titles
    car_titles.append(titles.text)#appending values
print(car_titles)#printing titles
len(car_titles)#printing length of list
```

```python
#Finding the description of cars using find_all
#All the description are present in dl class : spec-compact
car_description = []
a=soup.findAll('dl',{"class":"spec-compact"})
for mileage in a :
    print(mileage.text.strip())#printing description
    car_description.append(mileage.text.strip())#appending values
print(car_description)#printing description
len(car_description)#printing length of list
```

```python
#Finding the mileage of cars using find_all
#All the mileage are present in dd class : srp-odometer
car_mileage = []
a = soup.find_all('dd',{"class":"srp-odometer"})
for i in a:
    print(i.text) #printing individual values
    car_mileage.append(i.text) #appending values
print(car_mileage) #printing list
len(car_mileage) #length of list
```

```python
#Finding the exterior of cars using find_all
#All the exterior  are present in dd class : srp-color-exterior
car_exterior = []
a = soup.find_all('dd',{"class":"srp-color-exterior"})
for i in a:
    print(i.text) #printing individual values
    car_exterior.append(i.text) #appending values
print(car_exterior) #printing list
len(car_exterior) #length of list
```

```python
#Finding the engine of cars using find_all
#All the titles are present in dd class : srp-engine-combined
engine = []
a = soup.find_all('dd',{"class":"srp-engine-combined"})
for i in a:
    print(i.text) #printing individual values
    engine.append(i.text) #appending values
print(engine)#printing list
len(engine) #length of list
```

```python
#Finding the transmission of cars using find_all
#All the transmission are present in dd class : srp-transmission
transmission = []
a = soup.find_all('dd',{"class":"srp-transmission"})
for i in a:
    print(i.text)#printing individual values
    transmission.append(i.text) #appending values
print(transmission) #printing list
len(transmission) #length of list
```

```python
#Finding the drive type of cars using find_all
#All the drive type are present in dd class : srp-driveType
drive_type = []
a = soup.find_all('dd',{"class":"srp-driveType"})
for i in a:
    print(i.text) #printing individual values
    drive_type.append(i.text) #appending values
print(drive_type) #printing list
len(drive_type)#length of list
```

```python
#Finding the stock of cars using find_all
#All the stock are present in dd class : srp-stock-number
stock = []
a = soup.find_all('dd',{"class":"srp-stock-number"})
for i in a:
    print(i.text) #printing individual values
    stock.append(i.text) #appending values
print(stock) #printing list
len(stock) #length of list
```

```python
#Finding the vin of cars using find_all
#All the vin are present in dd class : srp-vin
vin = []
a = soup.find_all('dd',{"class":"srp-vin"})
for i in a:
    print(i.text) #printing individual values
    vin.append(i.text) #appending values
print(vin) #printing list
len(vin) #length of list
```

```python
#Finding the car price using find_all
#All the car price are present in strong class : price _bpcolor
car_price = []
a=soup.findAll('strong',{"class":"price _bpcolor"})
for c in a :
    print(c.text.strip())#printing car price
    car_price.append(c.text.strip())#appending values
print(car_price)#printing car price
len(car_price)#printing length of list
```

```python
#creating dict of scrapped data
dict = {
    'Title' : car_titles ,
    'Mileage':car_mileage,
    'Price': car_price ,
    'Exterior': car_exterior,
    'Engine' : engine,
    'Transmission':transmission,
    'Driveline':drive_type,
    'Stock' : stock ,
    'VIN' : vin  
}
```

```python
#creating dataframe from dict
df = pd.DataFrame.from_dict(dict,orient = 'index').T
df
```

```python
df['Dealer'] = 'itsworththetrip'
```
