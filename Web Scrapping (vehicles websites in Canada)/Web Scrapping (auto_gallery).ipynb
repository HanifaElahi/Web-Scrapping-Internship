```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from selenium import webdriver
import time
from matplotlib import pyplot as plt
```

```python
urla = "https://www.autogalleryofwinnipeg.com/vehicles/used/?st=asking_price,asc&sc=used"
```

```python
driver=webdriver.Firefox(executable_path="C:/Users/DELL/Downloads/geckodriver.exe")
#set the value of executable path
driver.get(urla)
```

```python
page_source=driver.page_source
```

```python
#parsing the page
soup=BeautifulSoup(page_source,'html.parser')
```

```python
#All the data present in a page is present in class : mb-lg list-view col-lg-12
data=soup.findAll('div',{"class":"mb-lg list-view col-lg-12"})
#printing the extracted data of a single page
for info in data:
    print(info.text)
```

```python
#Finding the titles of cars using find_all
#All the titles are present in div class : vehicle-card__full-details
a = soup.find_all("div",{"class" :"vehicle-card__full-details"})
car_titles = []#taking an empty list

#printing all the separate titles using for loop
for i in a:
    title = i.find('h3',{"class":"vehicle-card__title mb-xs"}).text #digging down for titles
    print(title)# printing values
    car_titles.append(title)  #appending values in a separate empty list
print(car_titles)#printing car titles list
len(car_titles)#length of car titles list
```

```python
vin = []
mileage = []
condition = []
body_style = []
stock = []
# taking empty lists


spec= soup.find_all("ul",{"class": "detailed-specs"}) #All the specifications are inside ul class: detailed-specs 

for i in spec:
    newsoup = BeautifulSoup(str(i),'html.parser')#using parser for scrapping inside details
    lis = newsoup.find_all('li')#finding all li inside ul
    for li in lis:
        
        if 'VIN' in li.text: # if present 
            vin.append(li.text.replace("VIN:","")) # then append
        if 'Kilometres' in li.text:# if present
            mileage.append(li.text.replace("Kilometres:",""))# then append
        if 'Condition' in li.text:# if present
            condition.append(li.text.replace("Condition:",""))# then append# then append
        if 'Body Style' in li.text:# if present
            body_style.append(li.text.replace("Body Style:",""))# then append
        if 'Stock' in li.text:# if present
            stock.append(li.text.replace("Stock #:",""))#then append
            
#priniting individual lists
print(vin)
print(mileage)
print(condition)
print(body_style)
print(stock)
```

```python
#Finding the prices of cars using find_all
#All the prices are present in div class : vehicle-card__price-wrap
car_price = []
c = soup.findAll('div',{"class":"vehicle-card__price-wrap"})
for i in c :
    sp=i.findAll('div',{"class":"price-block__price price-block__price--lg price-block__price--dark"})#digging further
    for j in sp:
        s=j.find('span',{"class":"df aifs"}).text.split("df aifs")[0]#actual values are inside span class :df aifs
        print(s)#printing prices
        car_price.append(s)#appending in a list
print(car_price)# printing list
len(car_price)# length of list
```

```python
#Finding the discount on cars using find_all
#All the discount are present in div class : vehicle-card__price-wrap
discount = []
c = soup.findAll('div',{"class":"vehicle-card__price-wrap"})
for i in c :
    sp=i.findAll('div',{"class":"price-block__single price-block--expanded price-block__single--finance"})#digging further
    for j in sp:
        s=j.find('span',{"class":"df aifs"}).text.split("df aifs")[0]#actual values are inside span class :df aifs
        print(s)#printing prices
        discount.append(s)#appending in a list
print(discount)#printing list
len(discount)#length of list
```

```python
#Finding the duration using find_all
#All the duration are present in div class : vehicle-card__price-wrap
duration = []
c = soup.findAll('div',{"class":"vehicle-card__price-wrap"})
for i in c :
    sp=i.findAll('div',{"class":"price-block__single price-block--expanded price-block__single--finance"})#digging further
    for j in sp:
        s=j.find('span',{"class":"price-block__sub"}).text.replace("\t","")#actual values are inside span class :df aifs
        print(s)#printing prices
        duration.append(s.strip())#appending in a list
print(duration)#printing list
len(duration)#length of list
```

```python
#creating dict of scrapped data
dict = {
    'Title' : car_titles ,
    'Mileage' : mileage , 
    'Price ': car_price ,
    'VIN' : vin ,
    'Stock' : stock ,
    'Body Style' : body_style ,
    'Condition' : condition,
    'Finanace Price per bw': discount ,
    'Finance Duration':duration
}
```

```python
#creating dataframe from dict
df = pd.DataFrame.from_dict(dict,orient = 'index').T
df
```

```python
split_title = df["Title"].str.split(" ", n = 3, expand = True) 
# splitting Tile column 
```

```python
df["Make Year"]= split_title[0]  
```

```python
df["Make"]= split_title[1] 
```

```python
df["Model"]= split_title[2] 
```

```python
df.drop(columns =["Title"], inplace = True) 
```
