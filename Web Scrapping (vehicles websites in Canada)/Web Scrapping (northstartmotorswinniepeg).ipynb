```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from selenium import webdriver
import time
from matplotlib import pyplot as plt
```

```python
urla = input(("Enter url : "))
# "https://www.northstarmotorswinnipeg.com/used-cars/"
```

```python
driver=webdriver.Firefox(executable_path="C:/Users/DELL/Downloads/geckodriver.exe")
#set the value of executable path
driver.get(urla)
```

```python
page_source=driver.page_source
```

```python
#parsing the page
soup=BeautifulSoup(page_source,'html.parser')
```

```python
#All the data present in a page is present in class row
data=soup.findAll('div',{"class":"row"})
#printing the extracted data of a single page
for info in data:
    print(info.text)
```

```python
#Finding the titles of cars using find_all
#All the titles are present in h3 class : vehicleName
car_titles = []
a=soup.findAll('h3',{"class":"vehicleName"})
for i in a :
    print(i.text)#printing titles
    car_titles.append(i.text)#appending values
print(car_titles)#printing titles
len(car_titles)#printing length of list
```

```python
#Finding the mileage of cars using find_all
#All the mileage are present in div class : Kilometers
car_mileage = []
a=soup.findAll('div',{"class":"Kilometers"})
for i in a :
    print(i.text.strip())#printing mileage
    car_mileage.append(i.text.replace("                                                                                      ","").strip())#appending values
print(car_mileage)#printing mileage
len(car_mileage)#printing length of list
```

```python
#Finding the price of cars using find_all
#All the price are present in div class : priceVal
car_price = []
a=soup.findAll('div',{"class":"priceVal"})
for i in a :
    print(i.text)#printing price
    car_price.append(i.text)#appending values
print(car_price)#printing price
len(car_price)#printing length of list
```

```python
stock = []
engine = []
fuel_type = []
exterior = []
body_style = []
drive_type = []
doors = []
transmission = []
# taking empty lists

#All the specifications are inside div class: col-md-6
spec= soup.find_all("div",{"class": "col-md-6"})

for i in spec:
    newsoup = BeautifulSoup(str(i),'html.parser')#using parser for scrapping inside details
    lis = newsoup.find_all('div',{"class":"vehicleDetailsInfo"})#finding all div with class vehicleDetailsInfo
    for li in lis:
        if 'Stock' in li.text: # if present 
            stock.append(li.text.replace("\nStock #:","").strip()) # then append
        if 'Body Style' in li.text:# if present
            body_style.append(li.text.replace("\nBody Style:","").strip())# then append
        if 'Driveline' in li.text:# if present
            drive_type.append(li.text.replace("\nDriveline:","").strip())# then append
        if 'Engine' in li.text:# if present
            engine.append(li.text.replace("\nEngine:","").strip())# then append# then append
        if 'Fuel Type' in li.text:# if present
            fuel_type.append(li.text.replace("\nFuel Type:","").strip())#then append
        if 'Doors' in li.text:# if present
            doors.append(li.text.replace("\nDoors:","").strip())# then append
        if 'Exterior' in li.text:# if present
            exterior.append(li.text.replace("\nExterior:","").strip())# then append# then append
        if 'Transmission' in li.text:# if present
            transmission.append(li.text.replace("\nTransmission:","").strip())#then append
            
            
#priniting individual lists
print(stock)
print(engine)
print(fuel_type)
print(exterior)
print(body_style)
print(drive_type)
print(doors)
print(transmission)
```

```python
car_status = []#taking an empty list
a = soup.findAll('div',{"class":"is-certified"})#scrapping status
for i in a:
    print(i.text)
    car_status.append(i.text)#appending in a list
print(car_status)#printing list
```

```python
#creating dict of scrapped data
dict = {
    'Title' : car_titles,
    'Price' : car_price,
    'Status' : car_status,
    'Mileage' : car_mileage,
    'Stock' : stock,
    'Exterior':exterior,
    'Engine':engine,
    'Fuel Type':fuel_type,
    'Body Style':body_style,
    'Drive Type':drive_type,
    'Doors':doors,
    'Transmission':transmission
}
```

```python
df = pd.DataFrame.from_dict(dict,orient = 'index').T
df
```

```python
#Dataframe to csv
df.to_csv('northsatartmotors.csv')
```
