```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from selenium import webdriver
import time
from matplotlib import pyplot as plt
```

```python
urla =  "https://www.focushyundai.net/new-vehicles/"
```

```python
driver=webdriver.Firefox(executable_path="C:/Users/DELL/Downloads/geckodriver.exe")
#set the value of executable path
driver.get(urla)
```

```python
page_source=driver.page_source
```

```python
#parsing the page
soup=BeautifulSoup(page_source,'html.parser')
```

```python
#All the data present in a page is present in class row 
data=soup.findAll('div',{"class":"row"})
#printing the extracted data of a single page
for info in data:
    print(info.text)
```

```python
#Finding the titles of cars using find_all
#All the titles are present in h2 class : result-title
car_titles = []
a=soup.findAll('h2',{"class":"result-title"})
for i in a :
    print(i.text)
    car_titles.append(i.text)#appending values
print(car_titles)#printing titles
len(car_titles)#printing length of list
```

```python
#Finding the make_year of cars using find_all
#All the make_year are present in span class : title-top
make_year = []
a=soup.findAll('span',{"class":"title-top"})
for i in a :
    print(i.text)
    make_year.append(i.text)#appending values
print(make_year)#printing make years
len(make_year)#printing length of list
```

```python
exterior = []
interior = []
# taking empty lists

#All the specifications are inside ul class: vehicle-details
spec= soup.find_all("ul",{"class": "vehicle-details"})

for i in spec:
    newsoup = BeautifulSoup(str(i),'html.parser')#using parser for scrapping inside details
    lis = newsoup.find_all('li')#finding all li inside ul
    for li in lis:
        if 'Exterior' in li.text:#if present
            exterior.append(li.text.replace("Exterior: ",""))#then append
        if 'Interior' in li.text:#if present
            interior.append(li.text.replace("Interior: ",""))#then append
            
#priniting individual lists
print(exterior)
print(interior)
```

```python
#Finding the msrp of cars using find_all
#All the  msrp are present in div class : hit-price hit-price--extra"
msrp = []
c = soup.findAll('div',{"class":"hit-price hit-price--extra"})
for i in c :
    sp=i.find('span',{"class":"ashallow"})   #digging further
    print(sp.text)#printing msrp
    msrp.append(sp.text)#appending in a list
print(msrp)#printing list
len(msrp)#length of list
```

```python
#Finding the savings of cars using find_all
#All the  savings are present in div class : hit-price hit-price--savings
savings = []
c = soup.findAll('div',{"class":"hit-price hit-price--savings"})
for i in c :
    sp=i.find('span',{"class":"ashallow"})    #digging further
    print(sp.text)#printing savings
    savings.append(sp.text)#appending in a list
print(savings)#printing list
len(savings)#length of list
```

```python
#Finding the selling prices of cars using find_all
#All the  selling prices are present in div class : hit-price hit-price--primary
car_price = []
c = soup.findAll('div',{"class":"hit-price hit-price--primary"})
for i in c :
    sp=i.find('span',{"class":"ashallow"})    #digging further
    print(sp.text)#printing car price
    car_price.append(sp.text)#appending in a list
print(car_price)#printing list
len(car_price)#length of list
```

```python
#creating dict of scrapped data
dict = {
    'Title' : car_titles ,
    'Make year' : make_year , 
    'Exterior ': exterior ,
    'Interior' : interior ,
    'MSRP' : msrp,
    'Price' : car_price,
    'Savings' : savings
    
    
}
```

```python
#creating dataframe from dict
df = pd.DataFrame.from_dict(dict,orient = 'index').T
df
```

```python
#Dataframe to csv
df.to_csv('focushyundai.csv')
```
