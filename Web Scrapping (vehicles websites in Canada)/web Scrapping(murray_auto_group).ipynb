```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from selenium import webdriver
import time
from matplotlib import pyplot as plt
```

```python
urla = "https://www.murrayautogroup.ca/vehicles/new/?view=grid&sc=new"
```

```python
driver=webdriver.Firefox(executable_path="C:/Users/DELL/Downloads/geckodriver.exe")
#set the value of executable path
driver.get(urla)
```

```python
page_source=driver.page_source
```

```python
#parsing the page
soup=BeautifulSoup(page_source,'html.parser')
```

```python
#All the data present in a page is present in class: row grid-row
data = soup.findAll('div',{"class":"row grid-row"})
#printing the extracted data of a single page
for info in data:
    print(info.text)
```

```python
#Finding the titles of cars using find_all
#All the titles are present in div class : vehicle-card__details
car_titles = []
a=soup.findAll('div',{"class":"vehicle-card__details"})
for i in a :
    titles=i.find('h4',{"class":"vehicle-card__title"}).text.strip()#all the titles are inside h4 class
    print(titles)#printing titles
    car_titles.append(titles)#appending values
print(car_titles)#printing titles
len(car_titles)#printing length of list
```

```python
#Finding the location of cars using find_all
#All the location are present in div class : vehicle-card__details
car_location = []
b = soup.findAll('div',{"class":"vehicle-card__details"})
for i in b :
    location=i.find('p',{"class":"vehicle-card__location"}).text.strip() #locations are inside p class
    print(location)#printing locations
    car_location.append(location)#apppending inside a separate list
print(car_location)#printing list
len(car_location)#length of list
```

```python
#Finding the prices of cars using find_all
#All the prices are present in div class : price-block__single
car_prices = []
c = soup.findAll('div',{"class":"price-block__single"})
for i in c :
    prices=i.find('span',{"class":"df aifs"}).text.strip()#all the prices are inside span class
    print(prices)#printing prices
    car_prices.append(prices)#appending in a list
print(car_prices)#printing list
len(car_prices)#length of list
```

```python
Vinlist = []
Kmlist = []
condition = []
body_style = []
engine = []
# taking empty lists

#All the specifications are inside ul class: detailed-specs inline-view
spec= soup.find_all("ul",{"class": "detailed-specs inline-view"})

for i in spec:
    newsoup = BeautifulSoup(str(i),'html.parser')#using parser for scrapping inside details
    lis = newsoup.find_all('li')#finding all li inside ul
    for li in lis:
        if 'VIN' in li.text: # if present 
            Vinlist.append(li.text.replace("VIN:","")) # then append
        if 'Kilometres' in li.text:# if present
            Kmlist.append(li.text.replace("Kilometres:",""))# then append
        if 'Condition' in li.text:# if present
            condition.append(li.text.replace("Condition:",""))# then append# then append
        if 'Body Style' in li.text:# if present
            body_style.append(li.text.replace("Body Style:",""))# then append
        if 'Engine' in li.text:# if present
            engine.append(li.text.replace("Engine:",""))
            
#priniting individual lists
print(Vinlist)
print(Kmlist)
print(condition)
print(body_style)
print(engine)

```

```python
#creating dict of scrapped data
dict = {
    'Title' : car_titles ,
    'Mileage' : Kmlist , 
    'Location' : car_location ,  
    'Price': car_prices ,
    'VIN' : Vinlist ,
    'Engine' : engine ,
    'Body Style' : body_style ,
    'Condition' : condition
}
```

```python
#creating dataframe from dict
df = pd.DataFrame.from_dict(dict,orient = 'index').T
df
```

```python
df['Dealer'] = 'murrayautogroup'
```

```python
split_title = df["Title"].str.split(" ", n = 2, expand = True) 
# splitting Tile column 
```

```python
df["Make Year"]= split_title[0]
```

```python
df["Make"]= split_title[1]
```

```python
df["Model"]= split_title[2] 
```
