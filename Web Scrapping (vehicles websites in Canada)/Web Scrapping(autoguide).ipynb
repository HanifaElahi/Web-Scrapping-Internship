```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from selenium import webdriver
import time
from matplotlib import pyplot as plt
```

```python
urla = input(("Enter url : "))
# "https://www.autoguide.com/used-cars/used-cars.html"
```

```python
driver=webdriver.Firefox(executable_path="C:/Users/DELL/Downloads/geckodriver.exe")
#set the value of executable path
driver.get(urla)
```

```python
page_source=driver.page_source
```

```python
#parsing the page
soup=BeautifulSoup(page_source,'html.parser')
```

```python
#All the data present in a page is present in id container
data=soup.findAll('div',{"id":"container"})
#printing the extracted data of a single page
for info in data:
    print(info.text)
```

```python
#Finding the titles of cars using find_all
#All the titles are present in a class : title
car_titles = []
a=soup.findAll('a',{"class":"title"})
for i in a :
    print(i.text)#printing titles
    car_titles.append(i.text)#appending values
print(car_titles)#printing titles
len(car_titles)#printing length of list
```

```python
mileage = []
location = []
mpg = []
hwy = []
combined = []
#taking empty lists

#All the specifications are inside ul class: stats
spec= soup.find_all("ul",{"class": "stats"})

for i in spec:
    newsoup = BeautifulSoup(str(i),'html.parser')#using parser for scrapping inside details
    lis = newsoup.find_all('li')#finding all li inside ul
    for li in lis:
        
        if 'Mileage' in li.text: # if present 
            mileage.append(li.text.replace(" Mileage:","")) # then append
        if 'Location' in li.text:# if present
            location.append(li.text.replace("Location: ",""))# then append
        if 'MPG' in li.text:# if present
            mpg.append(li.text.replace("MPG: ",""))# then append
        if 'hwy' in li.text:# if present
            hwy.append(li.text.replace("hwy ",""))# then append# then append
        if 'combined' in li.text:# if present
            combined.append(li.text.replace("combined",""))#then append
          
        
#priniting individual lists
print(mileage)
print(location)
print(mpg)
print(hwy)
print(combined)
```

```python
#Finding the  price on cars using find_all
#All the  price  are present in strong class : price
car_price = []
c = soup.findAll('strong',{"class":"price"})
for i in c :
    print(i.text)
    car_price.append(i.text)#appending in a list
print(car_price)#printing list
len(car_price)#length of list
```

```python
#creating dict of scrapped data
dict = {
    'Title' : car_titles ,
    'Mileage' : mileage, 
    'Location': location ,
    'MPG' : mpg ,
    'hwy' : hwy,
    'combined' :combined,
    'Price' : car_price
}
```

```python
#creating dataframe from dict
df = pd.DataFrame.from_dict(dict,orient = 'index').T
df
```

```python
#Dataframe to csv
df.to_csv('autoguide.csv')
```
