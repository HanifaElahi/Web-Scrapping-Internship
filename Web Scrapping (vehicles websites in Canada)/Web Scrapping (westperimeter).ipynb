```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from selenium import webdriver
import time
from matplotlib import pyplot as plt
```

```python
urla = "https://www.westperimeter.com/vehicles/?fn=&pr=&mk=&md=&tr=&bs=&od=&yr=&v1=Passenger%20Vehicles"
```

```python
driver=webdriver.Firefox(executable_path="C:/Users/DELL/Downloads/geckodriver.exe")
#set the value of executable path
driver.get(urla)
```

```python
page_source=driver.page_source
```

```python
#parsing the page
soup=BeautifulSoup(page_source,'html.parser')
```

```python
#All the data present in a page is present in class row 
data=soup.findAll('div',{"class":"row"})
#printing the extracted data of a single page
for info in data:
    print(info.text)
```

```python
#Finding the titles of cars using find_all
#All the titles are present in h4 class : vehicle-card__title 
car_titles = []
a=soup.findAll('h4',{"class":"vehicle-card__title"})
for i in a :
    print(i.text)#printing titles
    car_titles.append(i.text)#appending values
print(car_titles)#printing titles
len(car_titles)#printing length of list
```

```python
Vinlist = []
mileage = []
exterior = []
body_style = []
stock = []
# taking empty lists

#All the specifications are inside ul class: detailed-specs 
spec= soup.find_all("ul",{"class": "detailed-specs"})

for i in spec:
    newsoup = BeautifulSoup(str(i),'html.parser')#using parser for scrapping inside details
    lis = newsoup.find_all('li')#finding all li inside ul
    for li in lis:
       
        if 'VIN' in li.text: # if present 
            Vinlist.append(li.text.replace("VIN:","")) # then append
        if 'Kilometres' in li.text:# if present
            mileage.append(li.text.replace("Kilometres: ",""))# then append
        if 'Body Style' in li.text:# if present
            body_style.append(li.text.replace("Body Style:",""))# then append
        if 'Exterior' in li.text:# if present
            exterior.append(li.text.replace("Exterior Colour:",""))# then append# then append
        if 'Stock' in li.text:# if present
            stock.append(li.text.replace("Stock #:",""))#then append
            
#priniting individual lists
print(Vinlist)
print(mileage)
print(body_style)
print(exterior)
print(stock)
```

```python
#Finding the initial discount on cars using find_all
#All the  initial discount  are present in div class :vehicle-card__price-wrap
car_price = []
c = soup.findAll('div',{"class":"vehicle-card__price-wrap"})
for i in c :
    sp=i.findAll('div',{"class":"price-block__price price-block__price--sm price-block__price--primary"})#digging further
    for j in sp:
        s=j.find('span',{"class":"df aifs"}).text.split("df aifs")[0]#actual values are inside span class :df aifs
        print(s)#printing initial prices
        car_price.append(s)#appending in a list
print(car_price)#printing list
len(car_price)#length oflist

 
```

```python
#Finding the discounted price on cars using find_all
#All the  discounted price  are present in div class : price-block__single price-block--expanded price-block__single--finance
price = []
c = soup.findAll('div',{"class":"price-block__single price-block__single--finance"})
for i in c :
    sp=i.findAll('div',{"class":"price-block__price price-block__price--sm price-block__price--dark"})#digging further
    for j in sp:
        s=j.find('span',{"class":"df aifs"}).text.split("df aifs")[0]#actual values are inside span class :df aifs
        print(s)#printing discounted prices
        price.append(s)#appending in a list
print(price)#printing list
len(price)#length oflist

```

```python
#Finding the duration using find_all
#All the duration are present in div class : vehicle-card__price-wrap
duration = []
c = soup.findAll('div',{"class":"vehicle-card__price-wrap vehicle-card__price-wrap--finance"})
for i in c :
    s=i.find('span',{"class":"price-block__sub"})#actual values are inside span class :df aifs
    print(s.text.strip())#printing prices
    duration.append(s.text.strip())#appending in a list
print(duration)#printing list
len(duration)#length oflist
```

```python
#creating dict of scrapped data
dict = {
    'Title' : car_titles ,
    'Mileage' : mileage, 
    'Body Style': body_style ,
    'VIN' : Vinlist ,
    'Stock' : stock ,
    'Exterior' : exterior ,
    'Price' : price,
    'Finance Payment/bw': car_price,
    'Finance Payment Duration':duration
}
```

```python
df = pd.DataFrame.from_dict(dict,orient = 'index').T
df
```

```python
df['Dealer'] = 'westperimeter'
df['Condition'] = 'Used'
```

```python
split_title = df["Title"].str.split(" ", n = 3, expand = True) 
# splitting Tile column 
df["Make Year"]= split_title[0] 
df["Make"]= split_title[1] 
df["Model"]= split_title[2] 
df
```

```python
df.drop(columns =["Title"], inplace = True)
df
```

```python
#Dataframe to csv
df.to_csv('westperimeter.csv')
```
