```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from selenium import webdriver
import time
from matplotlib import pyplot as plt
```

```python
urla = "https://www.ridetime.ca/buy-used-cars/"
```

```python
driver=webdriver.Firefox(executable_path="C:/Users/DELL/Downloads/geckodriver.exe")
#set the value of executable path
driver.get(urla)
```

```python
page_source=driver.page_source
```

```python
#parsing the page
soup=BeautifulSoup(page_source,'html.parser')
```

```python
#All the data present in a page is present in class ninecol right last main__content-wrapper results results--products
data=soup.findAll('div',{"class":"ninecol right last main__content-wrapper results results--products"})
#printing the extracted data of a single page
for info in data:
    print(info.text)
```

```python
#Finding the titles of cars using find_all
#All the titles are present in h4 itemprop : name
car_titles = []
a=soup.findAll('h4',{"itemprop":"name"})
for i in a :
    car_titles.append(i.text.strip())#appending values
print(car_titles)#printing titles
len(car_titles)#printing length of list
```

```python
#Finding the stock no of cars using find_all
#All the stock no are present in p class : product-item__sku
stock = []
a=soup.findAll('p',{"class":"product-item__sku"})
for i in a :
    stock.append(i.text.replace("Stock: ","").strip())#appending values
print(stock)#printing stock
len(stock)#printing length of list
```

```python
#Finding the details of cars using find_all
#All the details are present in p class : product-item__description
details = []
a=soup.findAll('div',{"class":"product-item__description"})
for i in a :
    details.append(i.text.strip())#appending values
print(details)#printing details
len(details)#printing length of list
```

```python
#Finding the msrp of cars using find_all
#All the  msrp are present in h5 class : msrp
msrp = []
c = soup.findAll('h5',{"class":"msrp"})
for i in c :
    print(i.text.strip())
    msrp.append(i.text.replace("\n\t\t\t\t\tOriginal MSRP: ","").strip())#appending in a list
print(msrp)#printing list
len(msrp)#length oflist
```

```python
#Finding the sales price of cars using find_all
#All the  sales price are present in h5 class : full
car_price = []
c = soup.findAll('h4',{"class":"full"})
for i in c :
    print(i.text.strip())
    car_price.append(i.text.replace("Total Price:\n\t\t\t\t\t","").strip())#appending in a list
print(car_price)#printing list
len(car_price)#length oflist
```

```python
#Finding the finance price of cars using find_all
#All the  finance price are present in h5 class : bi-weekly
price = []
c = soup.findAll('h4',{"class":"bi-weekly"})
for i in c :
    print(i.text.strip())
    price.append(i.text.replace("Bi-Weekly:\n\t\t\t\t\t*","").strip())#appending in a list
print(price)#printing list
len(price)#length oflist
```

```python
#creating dict of scrapped data
dict = {
    'Title' : car_titles ,
    'Details':details,
    'Stock' : stock , 
    'MSRP' : msrp,
    'Price' : car_price,
    'Finance Payment /bw': price
}
```

```python
#creating dataframe from dict
df = pd.DataFrame.from_dict(dict,orient = 'index').T
df
```

```python
df['Condition'] = 'Used'
df['Dealer'] = 'ridetime'
```

```python
split_title = df["Title"].str.split(" ", n = 4, expand = True) 
# splitting Tile column
```

```python
df["Make Year"]= split_title[4]
```

```python
df["Make"]= split_title[0]
```

```python
df["Model"]= split_title[2]
```

```python
df.drop(columns =["Title"], inplace = True)
```
