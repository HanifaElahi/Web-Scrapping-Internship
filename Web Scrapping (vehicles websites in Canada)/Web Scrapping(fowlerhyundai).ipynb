```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from selenium import webdriver
import time
from matplotlib import pyplot as plt
```

```python
urla = "https://www.fowlerhyundai.com/vehicles/used/?st=model,asc&view=grid&sc=used"
```

```python
driver=webdriver.Firefox(executable_path="C:/Users/DELL/Downloads/geckodriver.exe")
#set the value of executable path
driver.get(urla)
```

```python
page_source=driver.page_source
```

```python
#parsing the page
soup=BeautifulSoup(page_source,'html.parser')
```

```python
#All the data present in a page is present in class row gird-row
data=soup.findAll('div',{"class":"row grid-row"})
#printing the extracted data of a single page
for info in data:
    print(info.text)
```

```python
#Finding the titles of cars using find_all
#All the titles are present in h4 class : vehicle-card__title
car_titles = []
a=soup.findAll('h3',{"class":"vehicle-card__title mb-xs"})
for i in a :
    print(i.text)
    car_titles.append(i.text)#appending values
print(car_titles)#printing titles
len(car_titles)#printing length of list
```

```python
vin = []
mileage = []
condition = []
stock = []
body_style = []
# taking empty lists

#All the specifications are inside ul class: detailed-specs
spec= soup.find_all("ul",{"class": "detailed-specs"})

for i in spec:
    newsoup = BeautifulSoup(str(i),'html.parser')#using parser for scrapping inside details
    lis = newsoup.find_all('li')#finding all li inside ul
    for li in lis:
        if 'VIN' in li.text: #if present
            vin.append(li.text.replace("VIN: ",""))#then append
        if 'Kilometres' in li.text:#if present
            mileage.append(li.text.replace("Kilometres: ",""))#then append
        if 'Condition' in li.text:#if present
            condition.append(li.text.replace("Condition: ",""))#then append
        if 'Body Style' in li.text:#if present
            body_style.append(li.text.replace("Body Style: ",""))#then append
        if 'Stock' in li.text:#if present
            stock.append(li.text.replace("Stock #:",""))#then append
        
#printing individual lists        
print(vin)
print(mileage)
print(condition)
print(body_style)
print(stock)
```

```python
#Finding the car price of cars using find_all
#All the  car price are present in div class : price-block__price price-block__price--sm price-block__price--dark
car_price = []
c = soup.findAll('div',{"class":"price-block__price price-block__price--lg price-block__price--dark"})
for i in c :
    sp=i.find('span',{"class":"df aifs"}) #digging further
    print(sp.text)#printing car price
    car_price.append(sp.text)#appending in a list
print(car_price)#printing list
len(car_price)#length of lis
```

```python
#Finding the finance payment of cars using find_all
#All the  finance payment  are present in div class : price-block__price price-block__price--sm price-block__price----primary
fn_price = []
c = soup.findAll('div',{"class":"price-block__price price-block__price--sm price-block__price--primary"})
for i in c :
    sp=i.find('span',{"class":"df aifs"})   #digging further
    print(sp.text)#printing finance payment 
    fn_price.append(sp.text)#appending in a list
print(fn_price)#printing list
len(fn_price)#length of list
```

```python
duration = [] #taking empty list
c = soup.findAll('div',{"class":"price-block__single price-block--expanded price-block__single--finance"}) #all the duration are inside div class vehicle-card__price-wrap vehicle-card__price-wrap--finance
for i in c :
    s = i.find('span',{"class":"price-block__sub"}) #digging further
    print(s.text.strip())
    duration.append(s.text.strip()) #appending in list
print(duration) #prinitng list
len(duration) #length of list
```

```python
#creating dict of scrapped data
dict = {
    'Title' : car_titles ,
    'VIN' : vin , 
    'Mileage': mileage,
    'Condition' : condition,
    'Stock' : stock,
    'Body Style' : body_style,
    'Price' : car_price,
    'Finance Payment/bw': fn_price,
    'Finance Payment Duration':duration
}
```

```python
#creating dataframe from dict
df = pd.DataFrame.from_dict(dict,orient = 'index').T
df
```

```python
df['Dealer'] = 'fowlerhyundai'
```

```python
split_title = df["Title"].str.split(" ", n = 2, expand = True) 
# splitting Tile column 
```

```python
df["Make Year"]= split_title[0]  
```

```python
df["Make"]= split_title[1] 
```

```python
df["Model"]= split_title[2] 
```

```python
df.drop(columns =["Title"], inplace = True)
```
