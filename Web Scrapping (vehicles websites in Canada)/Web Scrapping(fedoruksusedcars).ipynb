```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from selenium import webdriver
import time
from matplotlib import pyplot as plt
```

```python
urla = "https://www.fedoruksusedcars.ca/inventory/vehicle_type/car"
```

```python
driver=webdriver.Firefox(executable_path="C:/Users/DELL/Downloads/geckodriver.exe")
#set the value of executable path
driver.get(urla)
```

```python
page_source=driver.page_source
```

```python
#parsing the page
soup=BeautifulSoup(page_source,'html.parser')
```

```python
#All the data present in a page is present in class row
data=soup.findAll('div',{"class":"row"})
#printing the extracted data of a single page
for info in data:
    print(info.text)
```

```python
#Finding the titles of cars using find_all
#All the titles are present in li class : title
car_titles = []
a=soup.findAll('li',{"class":"title"})
for titles in a :
    print(titles.text)#printing titles
    car_titles.append(titles.text)#appending values
print(car_titles)#printing titles
len(car_titles)#printing length of list
```

```python
#Finding the mileage of cars using find_all
#All the mileage are present in span class : text-right
car_mileage = []
a=soup.findAll('span',{"class":"text-right"})
for mileage in a :
    print(mileage.text.strip())#printing mileage
    car_mileage.append(mileage.text.replace("                                                                                      ","").strip())#appending values
print(car_mileage)#printing mileage
len(car_mileage)#printing length of list
```

```python
#Finding the description using find_all
#All the description are present in li class : feature
car_description = []
a=soup.findAll('li',{"class":"feature"})
for cat in a :
    print(cat.text.strip())#printing description
    car_description.append(cat.text.replace("\n                              \n                               ","").strip())#appending values
print(car_description)#printing description
len(car_description)#printing length of list
```

```python
#Finding the car price using find_all
#All the car price are present in div class : price
car_price = []
a=soup.findAll('div',{"class":"price"})
for i in a :
    c = i.find('h4')
    car_price.append(c.text.strip())#appending values
print(car_price)#printing car price
len(car_price)#printing length of list
```

```python
#Finding the contact using find_all
#All the contact are present in div class : price
contact = []
a=soup.findAll('div',{"class":"price"})
for i in a:
    b=i.find('div',{"class":"phone"}) #digging further
    contact.append(b.text.strip())#appending values
print(contact)#printing contact
len(contact)#printing length of list
```

```python
#creating dict of scrapped data
dict = {
    'Title' : car_titles ,
    'Mileage' : car_mileage , 
    'Price': car_price ,
    'Description' : car_description,
    'Contact No.' : contact 
}
```

```python
#creating dataframe from dict
df = pd.DataFrame.from_dict(dict,orient = 'index').T
df
```

```python
df['Condition'] = 'Used'
df['Dealer'] = 'fedorukusedcars'
```

```python
split_title = df["Title"].str.split(" ", n = 2, expand = True) 
# splitting Tile column
```

```python
df["Make Year"]= split_title[0] 
```

```python
df["Make"]= split_title[1] 
```

```python
df["Model"]= split_title[2]
```

```python
df.drop(columns =["Title"], inplace = True) 
```
