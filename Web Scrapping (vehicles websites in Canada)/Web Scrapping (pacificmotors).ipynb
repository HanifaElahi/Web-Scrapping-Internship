```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from selenium import webdriver
import time
from matplotlib import pyplot as plt
```

```python
urla =  "https://www.pacificmotors.ca/used-cars-for-sale-in-winnipeg"
```

```python
driver=webdriver.Firefox(executable_path="C:/Users/DELL/Downloads/geckodriver.exe")
#set the value of executable path
driver.get(urla)
```

```python
page_source=driver.page_source
```

```python
#parsing the page
soup=BeautifulSoup(page_source,'html.parser')
```

```python
#All the data present in a page is present in class row inventory-list-row
data=soup.findAll('div',{"class":"row inventory-list-row"})
#printing the extracted data of a single page
for info in data:
    print(info.text)
```

```python
#Finding the titles of cars using find_all
#All the titles are present in div class : vehicle-label
car_titles = []
a=soup.findAll('div',{"class":"vehicle-label"})
for i in a :
    print(i.text.strip())#printing titles
    car_titles.append(i.text.replace("\n","").strip())#appending values
print(car_titles)#printing titles
len(car_titles)#printing length of list
```

```python
car_price = [] #taking empty list
a=soup.findAll('div',{"class":"pricevalue1 accent-color1"}) #prices are inside div class
for i in a :
    print(i.text)
    car_price.append(i.text)#appending in a list
print(car_price)#printing list
```

```python
exterior = []#taking empty list
a =soup.find_all('span',{"class":"Extcolor"})#scrapping
for i in a:
    print(i.text)
    exterior.append(i.text)#appending in a list
print(exterior)#printing list
len(exterior)#length of list
```

```python
transmission = []#taking empty list
a =soup.find_all('div',{"class":"transmission"})#scrapping
for i in a:
    print(i.text)
    transmission.append(i.text)#appending in a list
print(transmission)#printing list
len(transmission)#length of list
```

```python
fuel_type = []#taking empty list
a =soup.find_all('div',{"class":"fuel"})#scrapping
for i in a:
    print(i.text)
    fuel_type.append(i.text)#appending in a list
print(fuel_type)#printing list
len(fuel_type)#length of list
```

```python
engine = []#taking empty list
a =soup.find_all('div',{"class":"engine"})#scrapping
for i in a:
    print(i.text)
    engine.append(i.text)#appending in a list
print(engine)#printing list
len(engine)#length of list
```

```python
mileage = []#taking empty list
a =soup.find_all('span',{"class":"mileage"})#scrapping
for i in a:
    print(i.text)
    mileage.append(i.text)#appending in a list
print(mileage)#printing list
len(mileage)#length of list
```

```python
stock = []#taking empty list
a =soup.find_all('span',{"class":"stocknumber"})#scrapping
for i in a:
    print(i.text)
    stock.append(i.text)#appending in a list
print(stock)#printing list
len(stock)#length of list
```

```python
contact = []#taking empty list
a =soup.find_all('span',{"class":"phone"})#scrapping
for i in a:
    print(i.text)
    contact.append(i.text)#appending in a list
print(contact)#printing list
len(contact)#length of list
```

```python
vin = []#taking empty list
a =soup.find_all('span',{"class":"vin"})#scrapping
for i in a:
    print(i.text)
    vin.append(i.text.replace("VIN\n\n",""))#appending in a list
print(vin)#printing list
len(vin)#length of list
```

```python
location = []#taking empty list
a =soup.find_all('span',{"class":"mpg"})#scrapping
for i in a:
    print(i.text)
    location.append(i.text.replace("Location\n\n\n",""))#appending in a list
print(location)#printing list
len(location)#length of list
```

```python
#creating dict of scrapped data
dict = {
    'Title' : car_titles,
    'Price' : car_price,
    'Transmission':transmission,
    'Exterior':exterior,
    'Fuel Type':fuel_type,
    'Engine':engine,
    'Mileage': mileage,
    'Stock': stock,
    'VIN':vin,
    'Contact No.': contact,
    'Location':location
    
}
```

```python
df = pd.DataFrame.from_dict(dict,orient = 'index').T
df
```

```python
df['Dealer'] = 'pacificmotors'
df['Condition'] =  'Used'
```

```python
split_title = df["Title"].str.split(" ", n = 2, expand = True) 
# splitting Tile column 
df["Make Year"]= split_title[0] 
df["Make"]= split_title[1] 
df["Model"]= split_title[2] 
df
```

```python
df.drop(columns =["Title"], inplace = True)
df
```

```python
#Dataframe to csv
df.to_csv('pacificmotors.csv')
```
