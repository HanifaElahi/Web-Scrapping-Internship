```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from selenium import webdriver
import time
from matplotlib import pyplot as plt
```

```python
urla = input(("Enter url : "))
# "https://www.hondaonregent.ca/vehicles/?condition=used"
```

```python
driver=webdriver.Firefox(executable_path="C:/Users/DELL/Downloads/geckodriver.exe")
#set the value of executable path
driver.get(urla)
```

```python
page_source=driver.page_source
```

```python
#parsing the page
soup=BeautifulSoup(page_source,'html.parser')
```

```python
#All the data present in a page is present in class elementor-widget-container
data=soup.findAll('div',{"class":"elementor-widget-container"})
#printing the extracted data of a single page
for info in data:
    print(info.text)
```

```python
#Finding the titles of cars using find_all
#All the titles are present in span class : vehicle-year-make-model
car_titles = []
a=soup.findAll('span',{"class":"year-make-model"})
for i in a :
    print(i.text)#printing titles
    car_titles.append(i.text)#appending values
print(car_titles)#printing titles
len(car_titles)#printing length of list
```

```python
#Finding the details of cars using find_all
#All the details are present in span class : trim
details = soup.find_all("span",{"class" :"trim"})
#taking an empty list
car_details = []
#printing all the separate details using for loop
for d in details:
    print(d.text)# printing values
    #appending values in a separate empty list
    car_details.append(d.text)  
print(car_details)#printing car details list
len(car_details)#length of car details list
```

```python
#Finding the price of cars using find_all
#All the prices are present in span class : price
price_amount=soup.find_all("span",{"class" :"price"})
#taking an empty list
car_price=[]
#printing all the separate prices using for loop
for p in price_amount:
    print(p.text)# printing values
    #appending values in a separate empty list
    car_price.append(p.text)    
print(car_price)#printing car prices list
len(car_price)#length of car prices list
```

```python
#Finding the mileage of cars using find_all
#All the mileage values are present in span class : mileage
mileage=soup.find_all("span",{"class" :"mileage"})
#taking an empty list
car_mileage=[]
#printing all the separate mileage values using for loop
for m in mileage:
    
    print(m.text) # printing values
    #appending values in a separate empty list
    car_mileage.append(m.text.strip())    
print(car_mileage) #printing car mileage list
len(car_mileage) #length of car mileage list
```

```python
#Finding the condition of cars using find_all
#All the condition values are present in span class : condition 
condition = soup.find_all("span",{"class" :"condition"})
#taking an empty list
car_condition = []
#printing all the separate condition values using for loop
for c in condition:
    print(c.text) # printing values
    #appending values in a separate empty list
    car_condition.append(c.text)    
print(car_condition) #printing car condition list
len(car_condition) #length of car condition list
```

```python
#Creating dictionary of scrapped data(Car Title, Car Details, Price, Mileage, Condition)
#All these values are in individual lists
dict = {
        'Title' : car_titles , 
        'Details' : car_details , 
        'Price' : car_price , 
        'Mileage' : car_mileage, 
        'Condition': car_condition
       }
```

```python
df = pd.DataFrame.from_dict(dict,orient = 'index').T
df
```

```python
#Dataframe to csv
df.to_csv('handonregent_ca.csv')
```
