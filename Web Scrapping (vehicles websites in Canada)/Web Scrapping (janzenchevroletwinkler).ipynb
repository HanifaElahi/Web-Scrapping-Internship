```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from selenium import webdriver
import time
from matplotlib import pyplot as plt
```

```python
urla = input(("Enter url : "))
#"https://www.janzenchevroletwinkler.ca/VehicleSearchResults?search=new"
```

```python
driver=webdriver.Firefox(executable_path="C:/Users/DELL/Downloads/geckodriver.exe")
#set the value of executable path
driver.get(urla)
```

```python
page_source=driver.page_source
```

```python
#parsing the page
soup=BeautifulSoup(page_source,'html.parser')
```

```python
#All the data present in a page is present in class deck
data=soup.findAll('div',{"class":"deck"})
#printing the extracted data of a single page
for info in data:
    print(info.text)
```

```python
#Finding the titles of cars using find_all
#All the titles are present in h5 template : vehicle-name
car_titles = []
a=soup.findAll('h5',{"template":"vehicle-name"})
for i in a :
    print(i.text.strip())#printing titles
    car_titles.append(i.text.replace("\n","").strip())#appending values
print(car_titles)#printing titles
len(car_titles)#printing length of list
```

```python
engine = []
a =soup.find_all('li',{"template":"vehicleIdentitySpecs-engine"})
for i in a:
    print(i.text)
    engine.append(i.text.replace("Engine\n\n\n",""))
print(engine)
len(engine)
```

```python
interior = []
a =soup.find_all('span',{"itemprop":"vehicleInteriorColor"})
for i in a:
    print(i.text)
    interior.append(i.text)
print(interior)
len(interior)
```

```python
stock = []
a =soup.find_all('li',{"template":"vehicleIdentitySpecs-stockNumber"})
for i in a:
    print(i.text)
    stock.append(i.text.replace("Stock Number\n",""))
print(stock)
len(stock)
```

```python
vin = []
a =soup.find_all('li',{"template":"vehicleIdentitySpecs-vin"})
for i in a:
    print(i.text)
    vin.append(i.text.replace("VIN\n\n",""))
print(vin)
len(vin)
```

```python
transmission = []
a =soup.find_all('li',{"template":"vehicleIdentitySpecs-transmission"})
for i in a:
    print(i.text)
    transmission.append(i.text.replace("Transmission\n\n",""))
print(transmission)
len(transmission)
```

```python
exterior = []
a =soup.find_all('li',{"template":"vehicleIdentitySpecs-exterior"})
for i in a:
    print(i.text)
    exterior.append(i.text.replace("Exterior\n",""))
print(exterior)
len(exterior)
```

```python
cylinders = []
a =soup.find_all('li',{"template":"vehicleIdentitySpecs-cylinders"})
for i in a:
    print(i.text)
    cylinders.append(i.text.replace("Cylinders\n",""))
print(cylinders)
len(cylinders)
```

```python
#Finding the car old price  of cars using find_all
#All the car old price  are present in span class : value strikethrough
car_old_price = []
a=soup.findAll('span',{"class":"value strikethrough"})
for i in a :
    print(i.text)#printing old price
    car_old_price .append(i.text)#appending values
print(car_old_price)#printing old price
len(car_old_price)#printing length of list
```

```python
#Finding the price of cars using find_all
#All the price are present in span class: value big-and-bold
car_price = []
a=soup.findAll('span',{"class":"value big-and-bold"})
for i in a :
    print(i.text)#printing car price
    car_price.append(i.text)#appending values
print(car_price)#printing car price
len(car_price)#printing length of list
```

```python
#creating dict of scrapped data
dict = {
    'Title' : car_titles,
    'MSRP':car_old_price,
    'Price' : car_price,
    'Interior':interior,
    'Exterior':exterior,
    'Cylinders':cylinders,
    'Stock' : stock,
    'Engine':engine,
    'Transmission': transmission,
    'VIN':vin,
    'Transmission': transmission,

    
}
```

```python
df = pd.DataFrame.from_dict(dict,orient = 'index').T
df
```

```python
#Dataframe to csv
df.to_csv('janzenchevroletwinkler.csv')
```
