```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from selenium import webdriver
import time
from matplotlib import pyplot as plt
```

```python
urla = "https://www.carpages.ca/manitoba/winnipeg/used-cars/"
```

```python
driver=webdriver.Firefox(executable_path="C:/Users/DELL/Downloads/geckodriver.exe")
#set the value of executable path
driver.get(urla)
```

```python
page_source=driver.page_source
```

```python
#parsing the page
soup=BeautifulSoup(page_source,'html.parser')
```

```python
#All the data present in a page is present in class push-double--bottom
data=soup.findAll('div',{"class":"push-double--bottom"})
#printing the extracted data of a single page
for info in data:
    print(info.text)
```

```python
#Finding the titles of cars using find_all
#All the titles are present in div class : l-column l-column--large-8
a = soup.find_all("div",{"class" :"l-column l-column--large-8"})
#taking an empty list
car_title = []
#printing all the separate titles using for loop
for i in a:
    title = i.find('h4',{"class":"hN"}).text #diggig down for titles
    print(title)# printing values
    #appending values in a separate empty list
    car_title.append(title.strip())  
print(car_title)#printing car titles list
len(car_title)#length of car titles list
```

```python
#Finding the description of cars using find_all
#All the description  are present in div class : l-column l-column--large-8
a = soup.find_all("div",{"class" :"l-column l-column--large-8"})
#taking an empty list
car_description  = []
#printing all the separate description  using for loop
for i in a:
    description = i.find('h5',{"class":"hN grey"}).text #diggig down for description 
    print(description)# printing values
    #appending values in a separate empty list
    car_description.append(description)  
print(car_description)#printing car description  list
len(car_description )#length of car description list
```

```python
#Finding the dealer info using find_all
#All the dealer info are present in hgroup class : vehicle__card--dealerInfo
a = soup.find_all("hgroup",{"class" :"vehicle__card--dealerInfo"})
#taking an empty list
dealer_info = []
#printing all the separate dealer info using for loop
for i in a:
    b=i.find('h5',{"class": "hN"}).text
    c=i.find('p',{"class":"hN"}).text
    d = str(b) + str(c)
    print(d)
    #appending values in a separate empty list
    dealer_info.append(d)
print(dealer_info)#printing dealer info list
len(dealer_info)#length of dealer info list
```

```python
#Finding the prices of cars using find_all
#All the prices are present in div class : l-column l-column--medium-4 push-none
car_price = []
c = soup.findAll('div',{"class":"l-column l-column--medium-4 push-none"})
for i in c :
    print(i.text)
    car_price.append(i.text.replace("\n",""))#appending in a list
print(car_price)#printing list
len(car_price)#length oflist
```

```python
#Finding the mileage of cars using find_all
#All the mileage are present in div class : l-row soft-half--top
car_mileage = []
c = soup.findAll('div',{"class":"l-row soft-half--top"})
for i in c :
    b = i.find('div',{"class":"grey l-column l-column--small-6 l-column--medium-4"}) #digging further
    print(b.text)
    car_mileage.append(b.text)#appending in a list
print(car_mileage)#printing list
len(car_mileage)#length oflist

```

```python
#creating dict of scrapped data
dict = {
    'Title' : car_title ,
    'Description' : car_description , 
    'Price': car_price ,
    'Dealer' : dealer_info ,
    'Mileage' : car_mileage
}
```

```python
#creating dataframe from dict
df = pd.DataFrame.from_dict(dict,orient = 'index').T
df
```

```python
df['Condition'] = 'Used'
```

```python
split_title = df["Title"].str.split(" ", n = 2, expand = True) 
# splitting Tile column 
df["Make Year"]= split_title[0] 
df["Make"]= split_title[1] 
df["Model"]= split_title[2] 
df
```

```python
df.drop(columns =["Title"], inplace = True)
df
```

```python
#Dataframe to csv
df.to_csv('carpage_ca.csv')
```
