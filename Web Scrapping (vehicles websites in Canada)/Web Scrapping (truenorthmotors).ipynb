```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from selenium import webdriver
import time
from matplotlib import pyplot as plt
```

```python
urla = input(("Enter url : "))
# "https://www.truenorthmotors.ca/vehicles/"
```

```python
driver=webdriver.Firefox(executable_path="C:/Users/DELL/Downloads/geckodriver.exe")
#set the value of executable path
driver.get(urla)
```

```python
page_source=driver.page_source
```

```python
#parsing the page
soup=BeautifulSoup(page_source,'html.parser')
```

```python
#All the data present in a page is present in class global-container
data=soup.findAll('div',{"class":"global-container"})
#printing the extracted data of a single page
for info in data:
    print(info.text)
```

```python
#Finding the titles of cars using find_all
#All the titles are present in h4 class : epsilon weight--normal flush
car_titles = []
a=soup.findAll('h4',{"class":"epsilon weight--normal flush"})
for i in a :
    print(i.text)#printing titles
    car_titles.append(i.text.strip())#appending values
print(car_titles)#printing titles
len(car_titles)#printing length of list
```

```python
#Finding the mileage of cars using find_all
#All the titles are present in h4 class : zeta weight--normal
car_mileage = []
a=soup.findAll('h4',{"class":"zeta weight--normal"})
for i in a :
    print(i.text.strip())#printing mileage
    car_mileage.append(i.text.replace("Kilometers:\n              ","").strip())#appending values
print(car_mileage)#printing mileage
len(car_mileage)#printing length of list
```

```python
transmission = []
vin = []
location = []
body_style = []
engine = []
exterior = []


# taking empty lists

#All the specifications are inside div class: l-row
spec= soup.find_all("div",{"class": "l-row"})

for i in spec:
    newsoup = BeautifulSoup(str(i),'html.parser')#using parser for scrapping inside details
    lis = newsoup.find_all('tr')#finding all tr
    for li in lis:
        if 'Transmission' in li.text:# if present
            transmission.append(li.text.replace("Transmission\n","").strip())#then append
        if 'VIN' in li.text: # if present 
            vin.append(li.text.replace("\nVIN:","").strip()) # then append
        if 'Body Style' in li.text:# if present
            body_style.append(li.text.replace("\nBody Style:","").strip())# then append
        if 'City' in li.text:# if present
            location.append(li.text.replace("\nCity:","").strip())# then append
        if 'Engine' in li.text:# if present
            engine.append(li.text.replace("\nEngine:","").strip())# then append# then append
        if 'Exterior' in li.text:# if present
            exterior.append(li.text.replace("Exterior Colour:\n","").strip())# then append# then append
        
            
#priniting individual lists
print(transmission)
print(vin)
print(location)
print(body_style)
print(engine)
print(exterior)
```

```python
#Finding the price of cars using find_all
#All the price are present in span class : delta
car_price = []
a=soup.findAll('span',{"class":"delta"})
for i in a :
    print(i.text)#printing price
    car_price.append(i.text.strip())#appending values
print(car_price)#printing price
len(car_price)#printing length of list
```

```python
#Finding the stock of cars using find_all
#All the stock are present in p class : align--center weight--bold flush--bottom
stock = []
a=soup.findAll('p',{"class":"align--center weight--bold flush--bottom"})
for i in a :
    print(i.text)#printing stock
    stock.append(i.text.replace("Stock No:","").strip())#appending values
print(stock)#printing stock
len(stock)#printing length of list
```

```python
#creating dict of scrapped data
dict = {
    'Title' : car_titles,
    'Price' : car_price,
    'Mileage' : car_mileage,
    'Stock' : stock,
    'Exterior':exterior,
    'Engine':engine,
    'Body Style':body_style,
    'Transmission':transmission,
    'VIN':vin,
    'Location':location
}
```

```python
df = pd.DataFrame.from_dict(dict,orient = 'index').T
df
```

```python
#Dataframe to csv
df.to_csv('truenorthmotors.csv')
```
