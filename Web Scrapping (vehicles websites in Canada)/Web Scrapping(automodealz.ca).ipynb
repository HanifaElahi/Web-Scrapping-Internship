```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from selenium import webdriver
import time
from matplotlib import pyplot as plt
```

```python
urla = input(("Enter url : "))
# "https://www.automodealz.ca/used-cars"
```

```python
driver=webdriver.Firefox(executable_path="C:/Users/DELL/Downloads/geckodriver.exe")
#set the value of executable path
driver.get(urla)
```

```python
page_source=driver.page_source
```

```python
#parsing the page
soup=BeautifulSoup(page_source,'html.parser')
```

```python
#All the data present in a page is present in class container-inventory-page
data=soup.findAll('div',{"class":"container-inventory-page"})
#printing the extracted data of a single page
for info in data:
    print(info.text)
```

```python
#Finding the titles of cars using find_all
#All the titles are present in h3 class : vehicleName
car_titles = []
a=soup.findAll('h3',{"class":"vehicleName"})
for titles in a :
    print(titles.text)#printing titles
    car_titles.append(titles.text)#appending values
print(car_titles)#printing titles
len(car_titles)#printing length of list
```

```python
#Finding the mileage of cars using find_all
#All the mileage are present in div class : Kilometers
car_mileage = []
a=soup.findAll('div',{"class":"Kilometers"})
for mileage in a :
    print(mileage.text.strip())#printing mileage
    car_mileage.append(mileage.text.replace("                                                                                      ","").strip())#appending values
print(car_mileage)#printing mileage
len(car_mileage)#printing length of list
```

```python
stock=[]
body_style = []
engine = []
drive_line = []
fuel_type = []
doors = []
exterior = []
transmission = []
#taking individual empty lists

details = soup.find_all("div",{"class":"col-md-6"}) # all the details are inside div class:col-md-6
for i in details:
    if 'Stock #' in i.text: #if present
        stock.append(i.text.replace("Stock #:","").strip()) #then append
    if 'Body Style' in i.text:#if present
        body_style.append(i.text.replace("Body Style:","").strip())#then append
    if 'Engine' in i.text: #if present
        engine.append(i.text.replace("Engine:","").strip()) #then append
    if 'Driveline' in i.text:#if present
        drive_line.append(i.text.replace("Driveline:","").strip())#then append
    if 'Fuel Type' in i.text: #if present
        fuel_type.append(i.text.replace("Fuel Type:","").strip()) #then append
    if 'Doors' in i.text:#if present
        doors.append(i.text.replace("Doors:","").strip())#then append
    if 'Exterior' in i.text: #if present
        exterior.append(i.text.replace("Exterior:","").strip()) #then append
    if 'Transmission' in i.text:#if present
        transmission.append(i.text.replace("Transmission:","").strip())#then append

        
#printing lists
print(stock)
print(body_style)
print(engine)
print(drive_line)
print(fuel_type)
print(doors)
print(exterior)
print(transmission)
```

```python
#Finding the certification using find_all
#All the certification are present in div class : is-certified
car_certification = []
a=soup.findAll('div',{"class":"is-certified"})
for cat in a :
    print(cat.text.strip())#printing certification
    car_certification.append(cat.text)#appending values
print(car_certification)#printing certification
len(car_certification)#printing length of list
```

```python
#Finding the car price using find_all
#All the car price are present in div class : priceVal
car_price = []
a=soup.findAll('div',{"class":"priceVal"})
for c in a :
    print(c.text.strip())#printing car price
    car_price.append(c.text.strip())#appending values
print(car_price)#printing car price
len(car_price)#printing length of list
```

```python
#creating dict of scrapped data
dict = {
    'Title' : car_titles ,
    'Mileage' : car_mileage , 
    'Price ': car_price ,
    'Stock' : stock ,
    'Body Style' : body_style ,
    'Engine' : engine,
    'Driveline':drive_line,
    'Fuel Type':fuel_type,
    'Doors': doors,
    'Exterior': exterior,
    'Transmission':transmission,
    'Certification':car_certification
}
```

```python
#creating dataframe from dict
df = pd.DataFrame.from_dict(dict,orient = 'index').T
df
```

```python
#Dataframe to csv
df.to_csv('automodealz.csv')
```
