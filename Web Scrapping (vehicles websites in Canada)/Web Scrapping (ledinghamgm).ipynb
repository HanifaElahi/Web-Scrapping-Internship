```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from selenium import webdriver
import time
from matplotlib import pyplot as plt
```

```python
urla = input(("Enter url : "))
# "https://www.ledinghamgm.com/VehicleSearchResults?search=preowned"
```

```python
driver=webdriver.Firefox(executable_path="C:/Users/DELL/Downloads/geckodriver.exe")
#set the value of executable path
driver.get(urla)
```

```python
page_source=driver.page_source
```

```python
#parsing the page
soup=BeautifulSoup(page_source,'html.parser')
```

```python
#All the data present in a page is present in class deck
data=soup.findAll('div',{"class":"deck"})
#printing the extracted data of a single page
for info in data:
    print(info.text)
```

```python
#Finding the titles of cars using find_all
#All the titles are present in h4 template : vehicle-name
car_titles = []
a=soup.findAll('h4',{"template":"vehicle-name"})
for i in a :
    print(i.text)#printing titles
    car_titles.append(i.text.strip())#appending values
print(car_titles)#printing titles
len(car_titles)#printing length of list
```

```python
#Finding the price of cars using find_all
#All the price are present in span class : value big-and-bold
car_price = []
a=soup.findAll('span',{"class":"value big-and-bold"})
for i in a :
    print(i.text)#printing price
    car_price.append(i.text.strip())#appending values
print(car_price)#printing price
len(car_price)#printing length of list
```

```python
#Finding the msrp of cars using find_all
#All the msrp are present in span class : value strikethrough
msrp = []
a=soup.findAll('span',{"class":"value strikethrough"})
for i in a :
    print(i.text)#printing msrp
    msrp.append(i.text.strip())#appending values
print(msrp)#printing msrpe
len(msrp)#printing length of list
```

```python
#creating dict of scrapped data
dict = {
    'Title' : car_titles,
    'MSRP' : msrp,
    'Price' : car_price
}
```

```python
df = pd.DataFrame.from_dict(dict,orient = 'index').T
df
```

```python
#Dataframe to csv
df.to_csv('ledinghamgm.csv')
```
