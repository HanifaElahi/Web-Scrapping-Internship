```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from selenium import webdriver
import time
from matplotlib import pyplot as plt
```

```python
urla = "https://www.woodworthdodge.ca/new-inventory/index.htm"
```

```python
driver=webdriver.Firefox(executable_path="C:/Users/DELL/Downloads/geckodriver.exe")
#set the value of executable path
driver.get(urla)
```

```python
page_source=driver.page_source
```

```python
#parsing the page
soup=BeautifulSoup(page_source,'html.parser')
```

```python
#All the data present in a page is present in div class : page-section
data=soup.findAll('div',{"class":"page-section"})
#printing the extracted data of a single page
for info in data:
    print(info.text)
```

```python
#Finding the titles of cars using find_all
#All the titles are present in h2 class : vehicle-card-title mt-0 d-flex justify-content-between align-items-end
car_titles = []
a=soup.findAll('h2',{"class":"vehicle-card-title mt-0 d-flex justify-content-between align-items-end"})
for i in a :
    print(i.text)#printing titles
    car_titles.append(i.text)#appending values
print(car_titles)#printing titles
len(car_titles)#printing length of list
```

```python
engine = []#taking an empty list
a = soup.findAll('li',{"class":"engine"})#scrapping engine
for i in a:
    print(i.text)
    engine.append(i.text)#appending in a list
print(engine)#printing list
len(engine)#length of list
```

```python
transmission = []#taking an empty list
a = soup.findAll('li',{"class":"transmission"})#scrapping transmission
for i in a:
    print(i.text)
    transmission.append(i.text)#appending in a list
print(transmission)#printing list
len(transmission)#length of list
```

```python
exterior= []#taking an empty list
a = soup.findAll('li',{"class":"normalized-swatch-container exteriorColor"})#scrapping exterior
for i in a:
    print(i.text)
    exterior.append(i.text)#appending in a list
print(exterior)#printing list
len(exterior)#length of list
```

```python
interior= []#taking an empty list
a = soup.findAll('li',{"class":"normalized-swatch-container interiorColor"})#scrapping interior
for i in a:
    print(i.text)
    interior.append(i.text)#appending in a list
print(interior)#printing list
len(interior)#length of list
```

```python
stock = []#taking an empty list
a = soup.findAll('li',{"class":"stockNumber"})#scrapping stock
for i in a:
    print(i.text)
    stock.append(i.text.replace("Stock #:",""))#appending in a list
print(stock)#printing list
len(stock)#length of list
```

```python
model_code = []#taking an empty list
a = soup.findAll('li',{"class":"modelCode"})#scrapping model code
for i in a:
    print(i.text)
    model_code.append(i.text.replace("Model Code:",""))#appending in a list
print(model_code)#printing list
len(model_code)#length of list
```

```python
vin = []#taking an empty list
a = soup.findAll('li',{"class":"vin"})#scrapping vin
for i in a:
    print(i.text)
    vin.append(i.text)#appending in a list
print(vin)#printing list
len(vin)#length of list
```

```python
#Finding the msrp using find_all
#All the msrp are present in div class : msrp
msrp = []
a=soup.findAll('dd',{"class":"msrp"})
for c in a :
    print(c.text.strip())#printing msrp
    msrp.append(c.text.strip())#appending values
print(msrp)#printing msrp
len(msrp)#printing length of list
```

```python
#All the discount are present in dd class : discount text-success
discount = []
a=soup.findAll('dd',{"class":"discount text-success"})
for c in a :
    print(c.text.strip())#printing discount
    discount.append(c.text.strip())#appending values
print(discount)#printing discount
len(discount)#printing length of list
```

```python
#All the price  are present in dd class : final-price internetPrice font-weight-bold ddc-font-size-large line-height-reset pb-2
price = []
a=soup.findAll('dd',{"class":"final-price internetPrice font-weight-bold ddc-font-size-large line-height-reset pb-2"})
for c in a :
    print(c.text.strip())#printing price 
    price.append(c.text.strip())#appending values
print(price)#printing price 
len(price)#printing length of list
```

```python
#creating dict of scrapped data
dict = {
    'Title' : car_titles,
    'VIN' : vin, 
    'Exterior': exterior,
    'Interior':interior,
    'Model Code': model_code ,
    'Stock' : stock ,
    'Engine' : engine,
    'Transmission':transmission,
    'MSRP':msrp,
    'Discount':discount,
    'Price':price
}
```

```python
#creating dataframe from dict
df = pd.DataFrame.from_dict(dict,orient = 'index').T
df
```

```python
df['Condition'] = 'New'
df['Dealer'] = 'woodworthdodge'
```

```python
split_title = df["Title"].str.split(" ", n = 3, expand = True) 
# splitting Tile column 
```

```python
df["Make Year"]= split_title[0]  
```

```python
df["Make"]= split_title[1] 
```

```python
df["Model"]= split_title[3]
```

```python
df.drop(columns =["Title"], inplace = True) 
```
