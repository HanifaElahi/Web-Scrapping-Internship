```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from selenium import webdriver
import time
from matplotlib import pyplot as plt
```

```python
urla = "https://driverbase.com/inventory?search.CabTypeId=0&search.DealershipId=31992&search.DistanceFromZip=&search.EIdx=0&search.FindPriorityListings=false&search.InventoryType=&search.MakeId=0&search.MakeSlug=&search.MileageMaxString=&search.MileageMinString=&search.ModelId=0&search.ModelSlug=&search.MpgMaxString=&search.MpgMinString=&search.NewSearch=false&search.Next=false&search.Prev=false&search.PriceMaxString=&search.PriceMinString=&search.PropulsionTypeId=0&search.RequireImages=false&search.SIdx=0&search.SortTypeId=0&search.VehicleBodyTypeId=0&search.YearMax=0&search.YearMin=0&search.Zip="
```

```python
driver=webdriver.Firefox(executable_path="C:/Users/DELL/Downloads/geckodriver.exe")
#set the value of executable path
driver.get(urla)
```

```python
page_source=driver.page_source
```

```python
#parsing the page
soup=BeautifulSoup(page_source,'html.parser')
```

```python
#All the data present in a page is present in class col-sm-12
data=soup.findAll('div',{"class":"col-sm-12"})
#printing the extracted data of a single page
for info in data:
    print(info.text)
```

```python
#Finding the titles of cars using find_all
#All the titles are present in div class : valign-middle col-sm-4
car_titles = []
a=soup.findAll('td',{"class":"valign-middle col-sm-4"})
for i in a :
    titles = i.find('h2') #digging further
    print(titles.text)#printing titles
    car_titles.append(titles.text)#appending values
print(car_titles)#printing titles
len(car_titles)#printing length of list           
               
```

```python
car_certification = []
#taking empty list
a = soup.find_all('span',{"class":"label label-default square"}) # certifications are inside span class : label label-default square
for i in a:
    print(i.text)
    car_certification.append(i.text) #appending individual values
print(car_certification) #printing list
len(car_certification) #length of list
```

```python
car_price = [] #taking empty list

a = soup.find_all('h4',{"itemprop":"price"}) #all the prices are inside h4 itemprop : price

for i in a:
    print(i.text)
    car_price.append(i.text.strip()) #appending values 
print(car_price) #printing list
len(car_price) #length of list
```

```python
car_location =[] # taking empty list
a = soup.find_all('a',{"style":"padding-left:0px;padding-right:0px;"}) #all the locations are inside a style

for i in a:
    print(i.text)
    car_location.append(i.text) #appending values
print(car_location) #printing list
```

```python
car_mileage =[] #taking empty list
a = soup.find_all('h4',{"style":"padding-top:5px;padding-bottom:5px;color:gray;"}) # all the locations are inside h4 style 

for i in a:
    print(i.text)
    car_mileage.append(i.text)#appending values in a list

print(car_mileage) #printing list
```

```python
vin =  []
stock = []
mpg = []
cylinders = []
transmission = []
#taking empty lists

spec = soup.find_all('div',{"class":"row col-xs-6"}) # all the specifications are in div class : row col-xs-6
for i in spec:
    newsoup = BeautifulSoup(str(i),'html.parser')#using parser for scrapping inside details
    lis = newsoup.find_all('small')#finding all small tag inside div
    
    for li in lis:
        if 'VIN' in li.text:
            vin.append(li.text.replace("VIN: ",""))
        if 'Stock' in li.text:
            stock.append(li.text.replace("Stock: ",""))   
        if 'MPG' in li.text:
            mpg.append(li.text.replace("MPG:  ",""))   
        if 'Cylinders' in li.text:
            cylinders.append(li.text.replace("MPG:  ","")) 
        if 'WD' in li.text:
            transmission.append(li.text)

            
#printing individual lists            
print(vin)
print(stock)
print(mpg)
print(cylinders)
print(transmission)
```

```python
#creating dict of scrapped data
dict = {
    'Title' : car_titles,
    'Certification':car_certification,
    'Price':car_price,
    'VIN':vin,
    'Stock':stock,
    'MPG':mpg,
    'Cylinders':cylinders,
    'Location':car_location,
    'Mileage':car_mileage
}
```

```python
#creating dataframe from dict
df = pd.DataFrame.from_dict(dict,orient = 'index').T
df
```

```python
df['Condition'] = 'Used'
df['Dealer'] = 'driverbase'
```

```python
split_title = df["Title"].str.split(" ", n = 2, expand = True) 
# splitting Tile column 
df["Make Year"]= split_title[0] 
df["Make"]= split_title[1] 
df["Model"]= split_title[2] 
df
```

```python
df.drop(columns =["Title"], inplace = True)
df
```

```python
#Dataframe to csv
df.to_csv('driverbase.csv')
```
