```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from selenium import webdriver
import time
from matplotlib import pyplot as plt
```

```python
urla = "https://kelleherford.com/search/condition/used/"
```

```python
driver=webdriver.Firefox(executable_path="C:/Users/DELL/Downloads/geckodriver.exe")
#set the value of executable path
driver.get(urla)
```

```python
page_source=driver.page_source
```

```python
#parsing the page
soup=BeautifulSoup(page_source,'html.parser')
```

```python
#All the data present in a page is present in class flex flex-row flex-wrap
data=soup.findAll('div',{"class":"flex flex-row flex-wrap"})
#printing the extracted data of a single page
for info in data:
    print(info.text)
```

```python
#Finding the titles of cars using find_all
#All the titles are present in a class : title
a = soup.find_all("a",{"class" :"title"})
#taking an empty list
car_title = []
#printing all the separate titles using for loop
for i in a:
    print(i.text)
    car_title.append(i.text)  
print(car_title)#printing car titles list
len(car_title)#length of car titles list
```

```python
stock = [] 
#taking an empty list

#Finding stocks using find_all
#All the stocks are in p clasS:leading-none stock-number"
x = soup.find_all('p',{"class":"leading-none stock-number"})
for i in x:
    print(i.text.strip()) #printing text
    stock.append(i.text.replace("Stock #:","").strip()) #appending in an empty list
print(stock) #printing stock
len(stock) #printing length of list
```

```python
condition = [] 
#taking an empty list

#Finding condition using find_all
#All the condition are in span class: font-bold
x = soup.find_all('span',{"class":"font-bold"})
for i in x:
    print(i.text.strip()) #printing text
    condition.append(i.text) #appending in an empty list
print(condition) #printing condition
len(condition) #printing length of list
```

```python
engine = []
transmission = []
drive_train = []
exterior = []
mileage = []
# taking empty lists

#All the specifications are inside ul class: py-3 px-6 capitalize 
spec= soup.find_all("ul",{"class": "py-3 px-6 capitalize"})

for i in spec:
    newsoup = BeautifulSoup(str(i),'html.parser')#using parser for scrapping inside details
    lis = newsoup.find_all('li')#finding all li inside ul
    for li in lis:
        if 'Engine' in li.text: # if present 
            engine.append(li.text.replace("Engine:","")) # then append
        if 'Transmission' in li.text:# if present
            transmission .append(li.text.replace("Transmission:",""))# then append
        if 'Drivetrain' in li.text:# if present
            drive_train.append(li.text.replace("Drivetrain:",""))# then append# then append
        if 'Exterior' in li.text:# if present
            exterior.append(li.text.replace("Exterior:",""))# then append
        if 'Mileage' in li.text:# if present
            mileage .append(li.text.replace("Mileage:",""))#then append
            
#priniting individual lists
print(engine)
print(transmission)
print(drive_train)
print(exterior)
print(mileage)
```

```python
car_price = [] 
#taking an empty list

#Finding car price using find_all
#All the car price are in strong class: sale-price-value
x = soup.find_all('strong',{"class":"sale-price-value"})
for i in x:
    print(i.text.strip()) #printing text
    car_price.append(i.text) #appending in an empty list
print(car_price) #printing length
len(car_price) #printing length of list
```

```python
discount = [] 
#taking an empty list

#Finding discount using find_all
#All the discount are in div class: py-2 text-sm text-primary total-discounts
x = soup.find_all('div',{"class":"py-2 text-sm text-primary total-discounts"})
for i in x:
    print(i.text.strip()) #printing text
    discount.append(i.text.replace("YOU SAVE            ","").strip()) #appending in an empty list
print(discount) #printing discount
len(discount) #printing length of list
```

```python
#creating dict of scrapped data
dict = {
    'Title' : car_title ,
    'Stock' : stock ,
    'Engine' : engine , 
    'Transmission': transmission ,
    'Drive Train' : drive_train ,
    'Exterior' : exterior,
    'Mileage' : mileage ,
    'Condition' : condition,
    'Price': car_price,
    'Discount':discount
}
```

```python
#creating dataframe from dict
df = pd.DataFrame.from_dict(dict,orient = 'index').T
df
```

```python
df['Dealer'] = 'kellherford'
```

```python
split_title = df["Title"].str.split(" ", n = 3, expand = True) 
# splitting Tile column 
```

```python
df["Make Year"]= split_title[1]  
```

```python
df["Make"]= split_title[2] 
```

```python
df["Model"]= split_title[3]
```

```python
df.drop(columns =["Title"], inplace = True) 
```
