```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from selenium import webdriver
from matplotlib import pyplot as plt
```

```python
urla = "https://www.truecar.com/used-cars-for-sale/listings/tesla/"
```

```python
driver=webdriver.Firefox(executable_path="C:/Users/DELL/Downloads/geckodriver.exe")
#set the value of executable path
driver.get(urla)
```

```python
page_source=driver.page_source
```

```python
#parsing the page
soup=BeautifulSoup(page_source,'html.parser')
```

```python
#All the data present in a page is present in div class : row row-2 margin-bottom-3
data = soup.findAll('div',{"class":"row row-2 margin-bottom-3"})
#printing the extracted data of a single page
for info in data:
    print(info.text)
```

```python
#Finding the make years of cars using find_all
#All the make years are present in span class : vehicle-card-year font-size-1
make_year = soup.find_all("span",{"class" :"vehicle-card-year font-size-1"})

#taking an empty list
m_year = []
#printing all the separate years using for loop
for my in make_year:
    print(my.text)# printing values
    #appending values in a separate empty list
    m_year.append(my.text)  
print(m_year)#printing car making years list
len(m_year)#length of list
```

```python
#Finding the titles of cars using find_all
#All the titles are present in span class : vehicle-header-make-model text-truncate
titles = soup.find_all("span",{"class" :"vehicle-header-make-model text-truncate"})
```

```python
#taking an empty list
car_titles = []
#printing all the separate titles using for loop
for title in titles:
    print(title.text)# printing values
    #appending values in a separate empty list
    car_titles.append(title.text)  
print(car_titles)#printing car titles list
len(car_titles)#length of car titles list
```

```python
#Finding the major specifications of cars using find_all
#All the major specification are present in div data-test: vehicleCardTrim
major_spec = soup.find_all("div",{"data-test" :"vehicleCardTrim"})
```

```python
#taking an empty list
m_spec = []
#printing all the separate major specification using for loop
for ms in major_spec:
    print(ms.text)# printing values
    #appending values in a separate empty list
    m_spec.append(ms.text)  
print(m_spec)#printing car major specification list
len(m_spec)#length of car major specification list
```

```python
#Finding the mileage of cars using find_all
#All the mileage are present in div data-test : vehicleMileage
mileage = soup.find_all("div",{"data-test" :"vehicleMileage"})

#taking an empty list
car_mileage = []
#printing all the separate mileage using for loop
for m in mileage:
    print(m.text)# printing values
    #appending values in a separate empty list
    car_mileage.append(m.text)  
print(car_mileage)#printing car mileage list
len(car_mileage)#length of car mileage list
```

```python
#Finding the discount available on cars using find_all
#All the discount available are present in div class : padding-right-3 vehicle-card-bottom-pricing-primary
discount = soup.find_all("h4",{"class" :"heading-3 margin-y-1 font-weight-bold"})

#taking an empty list
car_discount = []
#printing all the separate discount available using for loop
for p in discount:
    print(p.text)# printing values
    #appending values in a separate empty list
    car_discount.append(p.text)  
print(car_discount)#printing car discount available list
len(car_discount)#length of car discount available list
```

```python
#Finding the location of cars using find_all
#All the location are present in div data-test : vehicleCardLocation
location = soup.find_all("div",{"data-test" :"vehicleCardLocation"})

#taking an empty list
car_location = []
#printing all the separate location using for loop
for l in location:
    print(l.text)# printing values
    #appending values in a separate empty list
    car_location.append(l.text)  
print(car_location)#printing car location list
len(car_location)#length of car location list
```

```python
#Finding the color of cars using find_all
#All the color are present in div data-test : vehicleCardColors
color = soup.find_all("div",{"data-test" :"vehicleCardColors"})

#taking an empty list
car_color = []
#printing all the separate color using for loop
for c in color:
    print(c.text)# printing values
    #appending values in a separate empty list
    car_color.append(c.text)  
print(car_color)#printing car color list
len(car_color)#length of car color list
```

```python
#Finding the condition of cars using find_all
#All the condition are present in div data-test : vehicleCardCondition
condition = soup.find_all("div",{"data-test" :"vehicleCardCondition"})

#taking an empty list
car_condition = []
#printing all the separate condition using for loop
for c in condition:
    print(c.text)# printing values
    #appending values in a separate empty list
    car_condition.append(c.text)  
print(car_condition)#printing car condition list
len(car_condition)#length of car condition list
```

```python
#Creating dictionary of scrapped data(Car Title, Make year, Specification, Mileage, Location, Discount, Color,Condition)
#All these values are in individual lists
dict = {
    'Title' : car_titles ,
    'Make Year' : m_year ,
    'Specification' : m_spec  ,
    'Mileage' : car_mileage , 
    'Discount' : car_discount , 
    'Location' : car_location , 
    'Color' : car_color, 
    'Condition History': car_condition
}
```

```python
df = pd.DataFrame.from_dict(dict,orient = 'index').T
df
```

```python
df['Condition'] = 'Used'
df['Dealer'] = 'truecar'
```

```python
split_title = df["Title"].str.split(" ", n = 1, expand = True) 
# splitting Tile column 
```

```python
df["Make"]= split_title[0]  
```

```python
df["Model"]= split_title[1] 
```

```python
df.drop(columns =["Title"], inplace = True)
```
