```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from selenium import webdriver
import time
from matplotlib import pyplot as plt
```

```python
urla = "http://www.westmanautos.ca/vehiclesearch.aspx?q=&cat_id=63&minprice=0&maxprice=99999&sort=item_posteddate%20DESC&tags=&make=&model=&year=&transmission=&kms=&vehicletype=Car"
```

```python
driver=webdriver.Firefox(executable_path="C:/Users/DELL/Downloads/geckodriver.exe")
#set the value of executable path
driver.get(urla)
```

```python
page_source=driver.page_source
```

```python
#parsing the page
soup=BeautifulSoup(page_source,'html.parser')
```

```python
#All the data present in a page is present in id maincontentbox
data=soup.findAll('div',{"id":"maincontentbox"})
#printing the extracted data of a single page
for info in data:
    print(info.text)
```

```python
#Finding the titles of cars using find_all
#All the titles are present in div id : listingitem_contentbox_title
car_titles = []
a=soup.findAll('div',{"id":"listingitem_contentbox_title"})
for i in a :
    print(i.text)# printing titles
    car_titles.append(i.text)# appending values
print(car_titles)# printing titles
len(car_titles)# printing length of list
```

```python
#Finding the car price using find_all
#All the car price are present in div id :listingitem_contentbox_price
car_price = []
a = soup.find_all('div',{"id":"listingitem_contentbox_price"})
for i in a : 
    print(i.text)#printing car price
    car_price.append(i.text)#appending values
print(car_price)#printing car price
len(car_price)#printing length of list
```

```python
#creating dict of scrapped data
dict = {
    'Title' : car_titles ,
    'Price': car_price 
}
```

```python
#creating dataframe from dict
df = pd.DataFrame.from_dict(dict,orient = 'index').T
df
```

```python
#Dataframe to csv
df.to_csv('westmanautos.csv')
```
