```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from selenium import webdriver
import time
from matplotlib import pyplot as plt
```

```python
urla = "https://www.dealerrater.ca/classifieds/search/Manitoba/?SelectedMakes=Ford&SelectedModels=F-150"
```

```python
driver=webdriver.Firefox(executable_path="C:/Users/DELL/Downloads/geckodriver.exe")
#set the value of executable path
driver.get(urla)
```

```python
page_source=driver.page_source
```

```python
#parsing the page
soup=BeautifulSoup(page_source,'html.parser')
```

```python
#All the data present in a page is present in id contentInner
data=soup.findAll('div',{"id":"contentInner"})
#printing the extracted data of a single page
for info in data:
    print(info.text)
```

```python
#Finding the titles of cars using find_all
#All the titles are present in h2 class : h2-header block line-height-1
car_titles = []
a=soup.findAll('h2',{"class":"h2-header block line-height-1"})
for i in a : 
    print(i.text.strip())#printing titles
    car_titles.append(i.text.strip())#appending values
print(car_titles)#printing titles
len(car_titles)#printing length of list
```

```python
#Finding the price of cars using find_all
#All the price are present in h2 class : h2-header block line-height-1
price = []
a=soup.findAll('a',{"class":"font-28 line-height-1 black boldest block margin-bottom-lg notranslate"})
for i in a : 
    print(i.text.strip())#printing price
    price.append(i.text.strip())#appending values
print(price)#printing price
len(price)#printing length of list
```

```python
color = []
transmission = []
#taking individual empty lists

details = soup.find_all("p",{"class":"margin-bottom-sm"}) # all the details are inside div class:shortDt
for i in details:

    if 'Color' in i.text:
            color.append(i.text.replace("Color:\n\n","").strip()) #then append
    if 'Transmission' in i.text:#if present
            transmission.append(i.text.replace("\nTransmission:\n","").strip())#then append         
#printing lists
print(color)
print(transmission)
```

```python
condition = []
a = soup.findAll('p',{"class":"bold uppercase font-13 letter-spacing-1 margin-bottom-none"})
for i in a:
    print(i.text)
    condition.append(i.text)
print(condition)
```

```python
#creating dict of scrapped data
dict = {
    'Title' : car_titles,
    'Color':color,
    'Transmission':transmission,
    'Price':price,
    'Condition':condition
}
```

```python
#creating dataframe from dict
df = pd.DataFrame.from_dict(dict,orient = 'index').T
df
```

```python
df['Dealer'] = 'dealerrater'
```

```python
split_title = df["Title"].str.split(" ", n = 2, expand = True) 
# splitting Tile column 
df["Make Year"]= split_title[0] 
df["Make"]= split_title[1] 
df["Model"]= split_title[2] 
df
```

```python
df.drop(columns =["Title"], inplace = True)
df
```

```python
#Dataframe to csv
df.to_csv('dealer_rater.csv')
```
