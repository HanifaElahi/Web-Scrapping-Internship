```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from selenium import webdriver
import time
from matplotlib import pyplot as plt
```

```python
urla = "https://www.crownmazda.ca/used-vehicles-in-winnipeg-mb/"
```

```python
driver=webdriver.Firefox(executable_path="C:/Users/DELL/Downloads/geckodriver.exe")
#set the value of executable path
driver.get(urla)
```

```python
page_source=driver.page_source
```

```python
#parsing the page
soup=BeautifulSoup(page_source,'html.parser')
```

```python
#All the data present in a page is present in td class result girdview
data = soup.findAll('td',{"class":"results gridview"})
#printing the extracted data of a single page
for info in data:
    print(info.text)
```

```python
#Finding the titles of cars using find_all
#All the titles are present in div class : vehicle-title
a = soup.find_all("div",{"class" :"vehicle-title"})
```

```python
#taking an empty list
car_titles = []
#printing all the separate titles using for loop
for i in a:
    titles = i.find('h2').text.split("listing")[0] # all titles are inside h2 
    print(titles.strip())# printing values
    #appending values in a separate empty list
    car_titles.append(titles.strip())  
print(car_titles)#printing car titles list
len(car_titles)#length of car titles list
```

```python
#Finding the price of cars using find_all
#All the price are present in div class : price-block our-price
c = soup.findAll("div",{"class" :"price-block our-price"})
#taking an empty list
car_price = []
#printing all the separate price using for loop
for i in c:
    prices = i.find("span",{"class" :"price"}).text.strip()
    print(prices)
        #print(p)# printing values
    #appending values in a separate empty list
    car_price.append(prices)  
print(car_price)#printing car price list
len(car_price)#length of car price list
```

```python
#Finding the location of cars using find_all
#All the location are present in div class : price-block bottom-text

car_location = []
#printing all the separate location using for loop
location = soup.findAll("div",{"class" :"price-block bottom-text"})
for i in location:
    print(i.text)
    #appending values in a separate empty list
    car_location.append(i.text.replace("Location:","").strip())  
print(car_location)#printing car location list
len(car_location)#length of car location list
```

```python
stock = []
#taking empty lists 

d = soup.find_all("div",{"class" :"vehInfo-vrp"}) #scrapping for  stock
for i in d:
    newsoup = BeautifulSoup(str(i),'html.parser')#using parser for scrapping inside details
    lis = newsoup.find_all('p')#finding p tag
    for li in lis: 
        if 'Stock' in li.text:
            stock.append(li.text.split('\n')[1].replace("Stock #: ",""))
            
print(stock)
```

```python
dict = {
    'Title' : car_titles,
    'Price' : car_price,
    'Location' : car_location,
    'Stock':stock
}
```

```python
df = pd.DataFrame.from_dict(dict,orient = 'index').T
```

```python
df
```

```python
split_title = df["Title"].str.split(" ", n = 4, expand = True) 
# splitting Tile column 
```

```python
df["Make Year"]= split_title[1]  
```

```python
df["Make"]= split_title[2] 
```

```python
df["Model"]= split_title[3] 
```

```python
df['Condition'] = 'Used'
```

```python
df.drop(columns =["Title"], inplace = True) 
```
