```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from selenium import webdriver
import time
from matplotlib import pyplot as plt
```

```python
urla = "https://www.dhillonauto.ca/active-inventory/"
```

```python
driver=webdriver.Firefox(executable_path="C:/Users/DELL/Downloads/geckodriver.exe")
#set the value of executable path
driver.get(urla)
```

```python
page_source=driver.page_source
```

```python
#parsing the page
soup=BeautifulSoup(page_source,'html.parser')
```

```python
#All the data present in a page is present in class row
data=soup.findAll('div',{"class":"row"})
#printing the extracted data of a single page
for info in data:
    print(info.text)
```

```python
#Finding the titles of cars using find_all
#All the titles are present in div class :title_holder
car_titles = []
a=soup.findAll('div',{"class":"title_holder"})
for i in a :
    x = i.find('h4',{"class":"name"}) #digging further for titles
    print(x.text.strip())#printing titles
    car_titles.append(x.text.replace("\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t   ","").strip())#appending values
print(car_titles)#printing titles
len(car_titles)#printing length of list
```

```python
vin=[]
stock = []
#taking individual empty lists

details = soup.find_all("ul",{"class":"options"}) # all the details are inside ul class: options
for i in details:
    newsoup = BeautifulSoup(str(i),'html.parser')#using parser for scrapping inside details
    lis = newsoup.find_all('li')#finding all li inside div ul
    for li in lis:
        if 'STOCK ID' in li.text: #if present
            stock.append(li.text.replace("STOCK ID : ","").strip()) #then append
        if 'VIN' in li.text:#if present
            vin.append(li.text.replace("VIN:","").strip())#then append

        
#printing lists
print(stock)
print(vin)
```

```python
mileage = []
doors = []
engine = []
#taking individual lists

a=soup.findAll('div',{"class":"col option col-12 col-md-12 col-lg-9"})
for i in a :
    newsoup = BeautifulSoup(str(i),'html.parser')#using parser for scrapping inside details
    lis = newsoup.find_all('li')#finding all li inside div class
    for li in lis:
        if 'Kms' in li.text: #if present
            mileage.append(li.text.replace("\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t","").strip()) #then append
        if 'Doors' in li.text:#if present
            doors.append(li.text.replace("\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t","").strip()) #then append
        if '0hp' in li.text:#if present
            engine.append(li.text.replace("\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t\t","").strip()) #then append

#printing lists
print(mileage)
print(doors)
print(engine)
```

```python
#Finding the car price using find_all
#All the car price are present in div class : priceVal
car_price = []
a=soup.findAll('span',{"class":"price"})
for c in a :
    print(c.text.strip())#printing car price
    car_price.append(c.text.strip())#appending values
print(car_price)#printing car price
len(car_price)#printing length of list
```

```python
#creating dict of scrapped data
dict = {
    'Title' : car_titles ,
    'VIN' : vin , 
    'Price': car_price ,
    'Stock' : stock ,
    'Engine' : engine,
    'Doors': doors,
    'Mileage':mileage
}
```

```python
#creating dataframe from dict
df = pd.DataFrame.from_dict(dict,orient = 'index').T
df
```

```python
df['Dealer'] = 'dhillonautos'
```

```python
split_title = df["Title"].str.split(" ", n = 2, expand = True) 
# splitting Tile column 
df["Make Year"]= split_title[0] 
df["Make"]= split_title[1] 
df["Model"]= split_title[2] 
df
```

```python
df.drop(columns =["Title"], inplace = True)
df
```

```python
#Dataframe to csv
df.to_csv('dhillonautos.csv')
```
