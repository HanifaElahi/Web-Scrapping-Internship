```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from selenium import webdriver
import time
from matplotlib import pyplot as plt
```

```python
urla = input(("Enter url : "))
#"https://www.harvesthonda.com/new/pg/10"
```

```python
driver=webdriver.Firefox(executable_path="C:/Users/DELL/Downloads/geckodriver.exe")
#set the value of executable path
driver.get(urla)
```

```python
page_source=driver.page_source
```

```python
#parsing the page
soup=BeautifulSoup(page_source,'html.parser')
```

```python
#All the data present in a page is present in class col-xs-12 col-sm-12 col-md-12
data=soup.findAll('div',{"class":"col-xs-12 col-sm-12 col-md-12"})
#printing the extracted data of a single page
for info in data:
    print(info.text)
```

```python
#Finding the titles of cars using find_all
#All the titles are present in p class : vehicle-year-make-model
car_titles = []
a=soup.findAll('p',{"class":"vehicle-year-make-model"})
for i in a :
    print(i.text)#printing titles
    car_titles.append(i.text)#appending values
print(car_titles)#printing titles
len(car_titles)#printing length of list
```

```python
body_style = []
exterior = []
transmission = []
stock = []
# taking empty lists

spec= soup.findAll('table',{"class":"table"}) #all the details are in table class

for i in spec:
    newsoup = BeautifulSoup(str(i),'html.parser')#using parser for scrapping inside details
    lis = newsoup.find_all('tr',{"class":"tr-even"})#finding all tr with class :  tr-even
    for li in lis:
        if 'Body Style' in li.text: #if present
            body_style.append(li.text.replace("Body Style:",""))#then append
        if 'Stock #' in li.text:#if present
            stock.append(li.text.replace("Stock #:",""))#then append
        if 'Transmission' in li.text:#if present
            transmission.append(li.text.replace("Transmission:",""))#then append
        if 'Exterior Colour' in li.text: #if present
            exterior.append(li.text.replace("Exterior Colour:",""))#then append
        
        
#printing individual lists        
print(body_style)
print(stock)
print(exterior)
print(transmission)
```

```python
engine = []
drive_train = []
location = []
# taking empty lists

spec= soup.findAll('table',{"class":"table"})# scrapping details

for i in spec:
    newsoup = BeautifulSoup(str(i),'html.parser')#using parser for scrapping inside details
    lis = newsoup.find_all('tr',{"class":"tr-odd"})#finding all tr-odd
    for li in lis:
        if 'Engine' in li.text: #if present
            engine.append(li.text.replace("Engine:",""))#then append
        if 'Drivetrain' in li.text:#if present
            drive_train.append(li.text.replace("Drivetrain:",""))#then append
        if 'City' in li.text:#if present
            location.append(li.text.replace("City:",""))#then append
        
        
#printing individual lists        
print(engine)
print(drive_train)
print(location)
```

```python
#Finding the car price on cars using find_all
#All the  car price  are present in span itemprop : price
car_price = []

c = soup.findAll('span',{"itemprop":"price"})
for i in c :
    print(i.text)
    car_price.append(i.text)
print(car_price)
```

```python
discount= []
lease = []
#taking individual lists

a = soup.findAll('p',{"class":"mileage-payment"})#scrapping details
for i in a:
    if 'Finance from' in i.text:#if present
        discount.append(i.text.replace("$34,285 x 84 Months @ 3.99% APR (estimated financing rate, cost of borrowing $5,037). \n$0 down payment. Taxes included.",""))#then append
    if 'Lease from' in i.text:#if present
        lease.append(i.text.replace("$34,285 x 60 Months @ 3.99% APR (estimated lease rate, cost of borrowing $5,625, estimated buyout $12,000). Taxes included.",""))#then append
#printing individual lists
print(discount)
print(lease)
```

```python
#creating dict of scrapped data
dict = {
    'Title' : car_titles,
    'Price' : car_price,
    'Body Style': body_style,
    'Exterior' : exterior ,
    'Stock' : stock,
    'Transmission' : transmission, 
    'Engine':engine,
    'Drive Train':drive_train,
    'Location':location,
    'Discount':discount,
    'Lease':lease
}
```

```python
df = pd.DataFrame.from_dict(dict,orient = 'index').T
df
```

```python
#Dataframe to csv
df.to_csv('harvesthonda.csv')
```
