```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from selenium import webdriver
import time
from matplotlib import pyplot as plt
```

```python
urla = input(("Enter url : ")) 
# "https://www.chuddschrysler.ca/new-inventory/index.htm?search=&compositeType=new&year=2020&make=Jeep&model=Cherokee&engine=&saveFacetState=true&lastFacetInteracted=inventory-listing1-facet-anchor-model-0"
```

```python
driver=webdriver.Firefox(executable_path="C:/Users/DELL/Downloads/geckodriver.exe")
#set the value of executable path
driver.get(urla)
```

```python
page_source=driver.page_source
```

```python
#parsing the page
soup=BeautifulSoup(page_source,'html.parser')
```

```python
#All the data present in a page is present in class bd
data=soup.findAll('div',{"class":"bd"})
#printing the extracted data of a single page
for info in data:
    print(info.text)
```

```python
#Finding the titles of cars using find_all
#All the titles are present in h3 class : fn
a = soup.find_all('h3',{"class":"fn"})
#taking an empty list
car_title = []
#printing all the separate titles using for loop
for i in a:
    print(i.text)# printing values
    #appending values in a separate empty list
    car_title.append(i.text)  
print(car_title)#printing car titles list
len(car_title)#length of car titles list
```

```python
engine = []
transmission = []
exterior = []
interior = []
stock = []
# taking empty lists

a = soup.find_all('div',{"class":"description"}) #all the details are inside div class : description
for i in a:
    e = i.find('dl').text.split(',')[0]#finding and splitting 
    engine.append(e.replace("Engine:",""))#appending
    f = i.find('dl').text.split(',')[1]#finding and splitting
    transmission.append(f.replace(" Transmission:",""))#appending
    g = i.find('dl').text.split(',')[2]#finding and splitting
    exterior.append(g.replace(" Exterior Colour:",""))#appending
    h = i.find('dl').text.split(',')[3]#finding and splitting
    interior.append(h.replace(" Interior Colour: ",""))#appending
    i = i.find('dl').text.split(',')[4]#finding and splitting
    stock.append(i.replace(" Stock #:",""))#appending

#printing individual lists     
print(engine)                 
print(transmission)
print(exterior)
print(interior)
print(stock)
```

```python
model_code = []
engine_size = []
fuel_type = []
# taking empty lists

spec = soup.find_all('div',{"class":"description"})#all the details are inside div class : description
for i in spec:
    a = i.find('dl',{"class":"last"}).text.split(',')[0]#finding and splitting
    model_code.append(a.replace("Model Code: ",""))#appending
    b = i.find('dl',{"class":"last"}).text.split(',')[1]#finding and splitting
    engine_size.append(b.replace(" Engine Size:",""))#appending
    c = i.find('dl',{"class":"last"}).text.split(',')[2]#finding and splitting
    fuel_type.append(c.replace("Fuel Type:",""))#appending

#printing individual lists   
print(model_code)
print(engine_size)
print(fuel_type)

```

```python
Vinlist = []#taking an empty list
a = soup.find_all('dl',{"class":"vin"})#scrapping vin
for i in a:
    print(i.text)
    Vinlist.append(i.text.replace("VIN: ",""))#appending in a list
print(Vinlist)#printing list
```

```python
msrp = []#taking an empty list
a = soup.find_all('span',{"class":"msrp"})#scrapping msrp
for i in a:
    print(i.text.strip())
    msrp.append(i.text.replace("MSRP: ","").strip())#appending in a list
print(msrp)
len(msrp)#printing list
```

```python
price_diff = []#taking an empty list
a = soup.find_all('span',{"class":"abSub"})#scrapping prce_diff
for i in a:
    print(i.text.strip())
    price_diff.append(i.text.replace("Price Difference:","").strip())#appending in a list
print(price_diff)
len(price_diff)#printing list
```

```python
finance = []#taking an empty list
b = soup.find_all('span',{"class":"paymentLoan"})#scrapping finance
for j in b:
    print(j.text.strip())
    finance.append(j.text.replace("\n\n","").strip())#appending in a list
print(finance)
len(finance)#printing list
```

```python
finance_dur = []#taking an empty list
b = soup.find_all('strong',{"class":"h3 pb-4 loanPaymentTermDescription align-right"})#scrapping finance_dur
for j in b:
    print(j.text.strip())
    finance_dur.append(j.text.replace("\n\n","").strip())#appending in a list
print(finance_dur)
len(finance_dur)#printing list
```

```python
lease = []#taking an empty list
b = soup.find_all('span',{"class":"paymentLease"})#scrapping lease
for j in b:
    print(j.text.strip())
    lease.append(j.text.replace("\n\n","").strip())#appending in a list
print(lease)
len(lease)#printing list
```

```python
lease_dur = []#taking an empty list
b = soup.find_all('strong',{"class":"h3 pb-4 leasePaymentTermDescription align-right"})#scrapping lease duration
for j in b:
    print(j.text.strip())
    lease_dur.append(j.text.replace("\n\n","").strip())#appending in a list
print(lease_dur)
len(lease_dur)#printing list
```

```python
retail_price = []#taking an empty list
a = soup.find_all('span',{"class":"internetPrice final-price"})#scrapping price
for i in a:
    print(i.text.strip())
    retail_price.append(i.text.replace("Retail Value:","").strip())#appending in a list
print(retail_price)
len(retail_price)#printing list
```

```python
#creating dict of scrapped data
dict = {
    'Title' : car_title,
    'Engine' : engine,
    'Transmission' : transmission,
    'Exterior':exterior,
    'Interior':interior,
    'Stock':stock,
    'Model Code':model_code,
    'Fuel Type':fuel_type,
    'Engine Size':engine_size,
    'VIN':Vinlist,
    'MSRP':msrp,
    'Price Difference':price_diff,
    'Finance Payment':finance,
    'Finance Duration':finance_dur,
    'Lease':lease,
    'Lease Duration':lease_dur,
    'Price':retail_price
}
```

```python
df = pd.DataFrame.from_dict(dict,orient = 'index').T
df
```

```python
#Dataframe to csv
df.to_csv('chuddchrysler.csv')
```
