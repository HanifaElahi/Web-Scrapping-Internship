```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from selenium import webdriver
import time
from matplotlib import pyplot as plt
```

```python
urla = "https://benrauto.com/inventory/"
```

```python
driver=webdriver.Firefox(executable_path="C:/Users/DELL/Downloads/geckodriver.exe")
#set the value of executable path
driver.get(urla)
```

```python
page_source=driver.page_source
```

```python
#parsing the page
soup=BeautifulSoup(page_source,'html.parser')
```

```python
#All the data present in a page is present in div class content-wrap car_listings row
data=soup.findAll('div',{"class":"content-wrap car_listings row"})
#printing the extracted data of a single page
for info in data:
    print(info.text)
```

```python
#Finding the titles of cars using find_all
#All the titles are present in div class : title
car_titles = []
a=soup.findAll('div',{"class":"title"})
for i in a :
    print(i.text)#printing titles
    car_titles.append(i.text)#appending values
print(car_titles)#printing titles
len(car_titles)#printing length of list
```

```python
exterior = []
interior = []
engine = []
mileage = []
vin = []
# taking empty lists

spec= soup.findAll('table',{"class":"options-secondary"}) # all th details are inside class : options-secondary

for i in spec:
    newsoup = BeautifulSoup(str(i),'html.parser')#using parser for scrapping inside details
    lis = newsoup.find_all('tr')#finding all tr inside table
    for li in lis:
        
        if 'Exterior' in li.text: #if present
            exterior.append(li.text.replace("Exterior Colour:",""))#then append
        if 'Interior' in li.text:#if present
            interior.append(li.text.replace("Interior Colour:",""))#then append
        if 'Engine Size' in li.text:#if present
            engine.append(li.text.replace("Engine Size: ",""))#then append
        if 'Kilometres' in li.text:#if present
            mileage.append(li.text.replace("Kilometres:",""))#then append
        if 'VIN' in li.text:#if present
            vin.append(li.text.replace("VIN: ",""))#then append
        
#printing individual lists        
print(exterior)
print(interior)
print(engine)
print(mileage)
print(vin)
```

```python
make = []
model = []
body_style = []
transmission = []

# taking empty lists

spec= soup.findAll('table',{"class":"options-primary"}) # all the details are inside class : options-primary

for i in spec:
    newsoup = BeautifulSoup(str(i),'html.parser')#using parser for scrapping inside details
    lis = newsoup.find_all('tr')#finding all tr inside table
    for li in lis:
        
        if 'Make' in li.text: #if present
            make.append(li.text.replace("Make: ",""))#then append
        if 'Model' in li.text:#if present
            model.append(li.text.replace("Model: ",""))#then append
        if 'Body Style' in li.text:#if present
            body_style.append(li.text.replace("Body Style: ",""))#then append
        if 'Transmission' in li.text:#if present
            transmission.append(li.text.replace("Transmission: ",""))#then append
        
#printing individual lists        
print(make)
print(model)
print(body_style)
print(transmission)
```

```python
#Finding the car price on cars using find_all
#All the  car price  are present in div class : figure
car_price = []
c = soup.findAll('div',{"class":"figure"})
for i in c :
    print(i.text)
    car_price.append(i.text.replace("\t\t\t\t\t\t\t",""))#appending in a list
print(car_price)#printing list
len(car_price)#length of list
```

```python
#creating dict of scrapped data
dict = {
    'Title' : car_titles ,
    'Exterior' : exterior , 
    'Interior ': interior,
    'VIN' : vin ,
    'Engine' : engine ,
    'Mileage': mileage,
    'Make' : make ,
    'Model' : model,
    'Body Style' : body_style,
    'Transmission' : transmission,
    'Price' : car_price
}
```

```python
df = pd.DataFrame.from_dict(dict,orient = 'index').T
df
```

```python
df['Condition'] = 'New'
```

```python
split_title = df["Title"].str.split(" ", n = 2, expand = True) 
# splitting Tile column 
```

```python
df["Make Year"]= split_title[0]  
```

```python
df.drop(columns =["Title"], inplace = True) 
```
