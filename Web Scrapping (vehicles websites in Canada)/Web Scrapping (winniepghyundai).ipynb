```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from selenium import webdriver
import time
from matplotlib import pyplot as plt
```

```python
urla = "https://www.winnipeghyundai.com/en/used-inventory"
```

```python
driver=webdriver.Firefox(executable_path="C:/Users/DELL/Downloads/geckodriver.exe")
#set the value of executable path
driver.get(urla)
```

```python
page_source=driver.page_source
```

```python
#parsing the page
soup=BeautifulSoup(page_source,'html.parser')
```

```python
#All the data present in a page is present in class inventory-listing-charlie
data=soup.findAll('section',{"class":"inventory-listing-charlie"})
#printing the extracted data of a single page
for info in data:
    print(info.text)
```

```python
make_year = []#taking an empty list
a = soup.findAll('span',{"class":"inventory-tile-section-vehicle-name--year-make"})#all the make years are inside span class : inventory-tile-section-vehicle-name--year-make
for i in a:
    print(i.text)
    make_year.append(i.text)#appending in a list
print(make_year)#printing list
len(make_year)#length of list
```

```python
#Finding the titles of cars using find_all
#All the titles are present in span class : inventory-tile-section-vehicle-name--model-name
car_titles = []
a=soup.findAll('span',{"class":"inventory-tile-section-vehicle-name--model-name"})
for i in a :
    print(i.text)
    car_titles.append(i.text)#appending values
print(car_titles)#printing titles
len(car_titles)#printing length of list
```

```python
stock = []#taking an empty list
a = soup.findAll('div',{"class":"inventory-tile-section-stock-number"})#all the stocks are inside div class : inventory-tile-section-stock-number
for i in a:
    print(i.text)
    stock.append(i.text.replace("\n# stock\n","").strip())#appending in a list
print(stock)#printing length
len(stock)#length of list
```

```python
msrp = []#taking an empty list
a = soup.findAll('div',{"class":"inventory-tile-section-price inventory-tile-section-price--section-list-price"})#scrapping msrp
for i in a:
    print(i.text)
    msrp.append(i.text.replace("\nRegular Price\n","").strip())#appending in a list
print(msrp)#printing list
```

```python
savings = []#taking an empty list
a = soup.findAll('span',{"class":"inventory-tile-section-price__price-difference"})#scrapping savings
for i in a:
    print(i.text)
    savings.append(i.text)#appending in a list
print(savings)#printing list
len(savings)#len of list
```

```python
#Finding the price of cars using find_all
#All the priceare present in strong class : price _bpcolor
car_price = []
a=soup.findAll('span',{"itemprop":"price"})
for i in a :
    print(i.text)
    car_price.append(i.text.strip())#appending values
print(car_price)#printing price
len(car_price)#printing length of list
```

```python
car_certification = []#taking empty list
a = soup.findAll('p',{"class":"text-bold inventory-tile-section-title-bar-block-new-used"})#scrapping description
for i in a:
    print(i.text)
    car_certification.append(i.text.strip())#appending in a list
print(car_certification)#printing list
len(car_certification)
```

```python
mileage = []#taking empty list
a = soup.findAll('div',{"itemprop":"mileageFromOdometer"})#scrapping mileage
for i in a:
    print(i.text)
    mileage.append(i.text.strip())#appending in a list
print(mileage)#printing list
```

```python
transmission = []#taking empty list
a = soup.findAll('div',{"itemprop":"vehicleTransmission"})#scrappping transmission
for i in a:
    print(i.text.strip())
    transmission.append(i.text.strip())#appending in a list
print(transmission)#printing list
```

```python
#creating a dict of scrapped data
dict = {
        'Title' : car_titles ,
        'Make Year':make_year,
        'Stock':stock,
        'MSRP':msrp,
        'Savings':savings,
        'Price':car_price,
        'Transmission':transmission,
        'Mileage':mileage,
        'Certification':car_certification
}
```

```python
df = pd.DataFrame.from_dict(dict,orient = 'index').T
df
```

```python
df['Dealer'] = 'winniepeghyundai'
```

```python
split_title = df["Title"].str.split(" ", n = 2, expand = True) 
# splitting Tile column 
df["Model"]= split_title[0] 
split_title = df["Make Year"].str.split(" ", n = 2, expand = True) 
# splitting Tile column 
df["Make Year"]= split_title[0] 
df["Make"]= split_title[1]
df
```

```python
df.drop(columns =["Title"], inplace = True)
df
```

```python
#Dataframe to csv
df.to_csv('winnipeghyuandai.csv')
```
