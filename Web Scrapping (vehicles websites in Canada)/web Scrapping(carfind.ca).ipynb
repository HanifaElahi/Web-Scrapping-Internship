```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from selenium import webdriver
import time
from matplotlib import pyplot as plt
```

```python
urla = "https://www.carfind.ca/en-CA/used/inventory/chevrolet/-/-/0"
```

```python
driver=webdriver.Firefox(executable_path="C:/Users/DELL/Downloads/geckodriver.exe")
#set the value of executable path
driver.get(urla)
```

```python
page_source=driver.page_source
```

```python
#parsing the page
soup=BeautifulSoup(page_source,'html.parser')
```

```python
#All the data present in a page is present in class listing
data=soup.findAll('div',{"class":"listing"})
#printing the extracted data of a single page
for info in data:
    print(info.text)
```

```python
#Finding the titles of cars using find_all
#All the titles are present in div class : vehicle
car_titles = []
a=soup.findAll('div',{"class":"vehicle"})
for i in a :
    print(i.text)#printing titles
    car_titles.append(i.text)#appending values
print(car_titles)#printing titles
len(car_titles)#printing length of list
```

```python
#Finding the description of cars using find_all
#All the description are present in div class : description
car_description = []
a=soup.findAll('div',{"class":"description"})
for i in a :
    print(i.text.strip())#printing description
    car_description.append(i.text)#appending values
print(car_description)#printing description
len(car_description)#printing length of list
```

```python
car_transmission = []
car_mileage = []
car_exterior = []
car_doors = []
car_condition = []
fuel_type = []
# taking empty lists

#All the specifications are inside div class: options
spec= soup.find_all("div",{"class": "options"})

for i in spec:
    newsoup = BeautifulSoup(str(i),'html.parser')#using parser for scrapping inside details
    lis = newsoup.find_all('li')#finding all li inside ul
    for li in lis:
        
        if 'Transmission' in li.text: # if present 
            car_transmission.append(li.text.replace("Transmission :","")) # then append
        if 'Mileage' in li.text:# if present
            car_mileage.append(li.text.replace("Mileage :",""))# then append
        if 'Exterior' in li.text:# if present
            car_exterior.append(li.text.replace("Exterior colour :",""))# then append
        if 'Doors' in li.text:# if present
            car_doors.append(li.text.replace("Doors :",""))# then append# then append
        if 'Condition' in li.text:# if present
            car_condition.append(li.text.replace("Condition :",""))#then append
        if 'Fuel Type' in li.text:# if present
            fuel_type.append(li.text.replace("Fuel Type :",""))#then append
            
#priniting individual lists
print(car_transmission)
print(car_mileage)
print(car_exterior)
print(car_doors)
print(car_condition)
print(fuel_type)
```

```python
#Finding the car price of cars using find_all
#All the  car price  are present in div class : col-sm-2 right
car_price = []
c = soup.findAll('div',{"class":"col-sm-2 right"})
for i in c :
    s=i.find('span',{"class":"price"}).text#actual values are inside span class : price
    print(s)#printing initial prices
    car_price.append(s)#appending in a list
print(car_price)#printing list
len(car_price)#length of list
```

```python
#Finding the stocks of cars using find_all
#All the  stocks  are present in div class : col-sm-2 right
stock = []
c = soup.findAll('div',{"class":"col-sm-2 right"})
for i in c :
    s=i.find('span',{"class":"stock-number"}).text.replace("stock # : ","")#actual values are inside span class :stock-number
    print(s)#printing stocks
    stock.append(s)#appending in a list
print(stock)#printing list
len(stock)#length of list
```

```python
#creating dict of scrapped data
dict = {
    'Title' : car_titles ,
    'Description' : car_description,
    'Transmission' : car_transmission , 
    'Mileage': car_mileage ,
    'Exterior' : car_exterior ,
    'Doors' : car_doors ,
    'Condition' : car_condition ,
    'Fuel Type' : fuel_type ,
    'Price' : car_price,
    'Stock' : stock
}
```

```python
#creating dataframe from dict
df = pd.DataFrame.from_dict(dict,orient = 'index').T
df
```

```python
df['Dealer'] = 'carfind'
```

```python
split_title = df["Title"].str.split(" ", n = 2, expand = True) 
# splitting Tile column 
df["Make Year"]= split_title[0] 
df["Make"]= split_title[1] 
df["Model"]= split_title[2] 
df
```

```python
df.drop(columns =["Title"], inplace = True)
df
```

```python
#Dataframe to csv
df.to_csv('carfind.csv')
```
