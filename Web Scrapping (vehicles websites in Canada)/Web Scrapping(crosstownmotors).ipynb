```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from selenium import webdriver
import time
from matplotlib import pyplot as plt
```

```python
urla = input(("Enter url : "))
# "https://www.crosstownmotors.com/en/new-inventory"
```

```python
driver=webdriver.Firefox(executable_path="C:/Users/DELL/Downloads/geckodriver.exe")
#set the value of executable path
driver.get(urla)
```

```python
page_source=driver.page_source
```

```python
#parsing the page
soup=BeautifulSoup(page_source,'html.parser')
```

```python
#All the data present in a page is present in class inventory-listing-charlie__content
data=soup.findAll('section',{"class":"inventory-listing-charlie__content"})
#printing the extracted data of a single page
for info in data:
    print(info.text)
```

```python
#Finding the titles of cars using find_all
#All the titles are present in div class : inventory-preview-bravo__infos-wrapper
car_titles = []
a=soup.findAll('div',{"class":"inventory-preview-bravo__infos-wrapper"})
for i in a :
    titles = i.find('span',{"itemprop":"name"}) #digging further
    print(titles.text)#printing titles
    car_titles.append(titles.text)#appending values
print(car_titles)#printing titles
len(car_titles)#printing length of list
```

```python
#Finding the stock of cars using find_all
#All the stock are present in p class : inventory-preview-bravo__stock-number
stock = []
a=soup.findAll('p',{"class":"inventory-preview-bravo__stock-number"})
for i in a :
    print(i.text.strip())#printing titles
    stock.append(i.text.replace("Stock # ","").strip())#appending values
print(stock)#printing titles
len(stock)#printing length of list
```

```python
transmission = []
drive_train = []
#taking individual empty lists

details = soup.find_all("div",{"class":"inventory-preview-bravo__options"}) # all the details are inside div class: inventory-preview-bravo__options
for i in details:
    newsoup = BeautifulSoup(str(i),'html.parser')#using parser for scrapping inside details
    for li in newsoup:
         
        if 'Transmission' in newsoup.text:#if present
            transmission.append(newsoup.text.replace("Transmission\n","").strip())#appending
        if 'Drivetrain' in newsoup.text:#if present
            drive_train.append(newsoup.text.replace("Drivetrain\n","").strip()) #appending   

#printing lists
print(transmission)
print(drive_train)
```

```python
car_price = []#taking empty lists
a=soup.findAll('span',{"class":"vehicle-cash-purchase-selling-price__value"})#scrapping car price
for i in a :
    print(i.text.strip())
    car_price.append(i.text.strip())#appending in a list
print(car_price)#printing list
```

```python
#Finding the finance payment using find_all
#All the finance payment are present in div class : leaseBox
finance_payment = []
a=soup.findAll('span',{"class":"vehicle-finance__item-term-payment-value"})
for c in a :
    print(c.text.strip())#printing finance payment
    finance_payment.append(c.text.strip())#appending values
print(finance_payment)#printing finance payment
len(finance_payment)#printing length of list
```

```python
#Finding the duration using find_all
#All the duration are present in div class : leaseBox
duration = []
a=soup.findAll('p',{"class":"vehicle-finance__item-value columns"})
for c in a :
    print(c.text.strip())#printing duration
    duration.append(c.text.replace("\n            ","").strip())#appending values
print(duration)#printing duration
len(duration)#printing length of list
```

```python
#creating dict of scrapped data
dict = {
    'Title' : car_titles,
    'Stock' : stock ,
    'Drive Train':drive_train,
    'Transmission':transmission,
    'Price':car_price,
    'Finance Payment':finance_payment
}
```

```python
#creating dataframe from dict
df = pd.DataFrame.from_dict(dict,orient = 'index').T
df
```

```python
#Dataframe to csv
df.to_csv('crosstownmotors.csv')
```
