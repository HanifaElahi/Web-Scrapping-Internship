```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from selenium import webdriver
import time
from matplotlib import pyplot as plt
```

```python
urla = "https://www.precisiontoyota.ca/new/"
```

```python
driver=webdriver.Firefox(executable_path="C:/Users/DELL/Downloads/geckodriver.exe")
#set the value of executable path
driver.get(urla)
```

```python
page_source=driver.page_source
```

```python
#parsing the page
soup=BeautifulSoup(page_source,'html.parser')
```

```python
#All the data present in a page is present in class row
data=soup.findAll('div',{"class":"row"})
#printing the extracted data of a single page
for info in data:
    print(info.text)
```

```python
#Finding the titles of cars using find_all
#All the titles are present in p class : vehicle-year-make-model
car_titles = []
a=soup.findAll('p',{"class":"vehicle-year-make-model"})
for i in a :
    print(i.text)#printing titles
    car_titles.append(i.text)#appending values
print(car_titles)#printing titles
len(car_titles)#printing length of list
```

```python
#Finding the description of cars using find_all
#All the descriptionare present in h2 class : vehicle-infoline vlp-extended-description
car_description = []
a=soup.findAll('h2',{"class":"vehicle-infoline vlp-extended-description"})
for i in a :
    print(i.text)#printing description
    car_description.append(i.text)#appending values
print(car_description)#printing description
len(car_description)#printing length of list
```

```python
#Finding the car  price   using find_all
#All the car price are present in span itemprop : price
car_price = []
a=soup.findAll('span',{"itemprop":"price"})
for i in a :
    print(i.text)#printing  price
    car_price .append(i.text)#appending values
print(car_price)#printing  price
len(car_price)#printing length of list
```

```python
transmission = []
body_style = []
exterior = []
stock = []
# taking empty lists

spec= soup.findAll('table',{"class":"table"})#all the specifications are inside table class
for i in spec:
    newsoup = BeautifulSoup(str(i),'html.parser')#using parser for scrapping inside details
    lis = newsoup.find_all('tr',{"class":"tr-even"})#finding all tr inside table wit class tr-even
    for li in lis:
        
        if 'Transmission' in li.text: #if present
            transmission.append(li.text.replace("Transmission:",""))#then append
        if 'Exterior Colour' in li.text:#if present
            exterior.append(li.text.replace("Exterior Colour:",""))#then append
        if 'Body Style' in li.text:#if present
            body_style.append(li.text.replace("Body Style:",""))#then append
        if 'Stock' in li.text: #if present
            stock.append(li.text.replace("Stock #:","")) #then append

            #printing individual lists        
print(transmission)
print(exterior)
print(body_style)
print(stock)
```

```python
engine = []
drive_train = []
location = []
# taking empty lists

spec= soup.findAll('table',{"class":"table"})#all the specifications are inside table class
for i in spec:
    newsoup = BeautifulSoup(str(i),'html.parser')#using parser for scrapping inside details
    lis = newsoup.find_all('tr',{"class":"tr-odd"})#finding all tr inside table wit class tr-odd
    for li in lis:
       
        if 'Engine' in li.text: #if present
            engine.append(li.text.replace("Engine:",""))#then append
        if 'Drivetrain' in li.text:#if present
            drive_train.append(li.text.replace("Drivetrain:",""))#then append
        if 'City' in li.text:#if present
            location.append(li.text.replace("City:",""))#then append
       
 #printing individual lists        
print(engine)
print(drive_train)
print(location)
```

```python
#creating dict of scrapped data
dict = {
    'Title' : car_titles,
    'Price' : car_price,
    'Transmission' : transmission,
    'Exterior':exterior,
    'Body Style':body_style,
    'Stock' : stock,
    'Engine' : engine ,
    'Drive Train':drive_train,
    'Location':location
}
```

```python
df = pd.DataFrame.from_dict(dict,orient = 'index').T
df
```

```python
df['Dealer'] = 'precisiontoyota'
df['Condition'] =  'New'
```

```python
split_title = df["Title"].str.split(" ", n = 2, expand = True) 
# splitting Tile column 
df["Make Year"]= split_title[0] 
df["Make"]= split_title[1] 
df["Model"]= split_title[2] 
df
```

```python
df.drop(columns =["Title"], inplace = True)
df
```

```python
#Dataframe to csv
df.to_csv('precisiontoyota.csv')
```
