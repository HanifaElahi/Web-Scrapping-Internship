```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from selenium import webdriver
import time
from matplotlib import pyplot as plt
```

```python
urla = input(("Enter url : "))
#"https://www.funkstoyota.ca/new-toyota-vehicles-south-of-winnipeg-in-steinbach"
```

```python
driver=webdriver.Firefox(executable_path="C:/Users/DELL/Downloads/geckodriver.exe")
#set the value of executable path
driver.get(urla)
```

```python
page_source=driver.page_source
```

```python
#parsing the page
soup=BeautifulSoup(page_source,'html.parser')
```

```python
#All the data present in a page is present in class row
data=soup.findAll('div',{"class":"row"})
#printing the extracted data of a single page
for info in data:
    print(info.text)
```

```python
#Finding the titles of cars using find_all
#All the titles are present in h4 class : vehicle-header media-heading
car_titles = []
a=soup.findAll('h4',{"class":"vehicle-header media-heading"})
for i in a :
    print(i.text)#printing titles
    car_titles.append(i.text)#appending values
print(car_titles)#printing titles
len(car_titles)#printing length of list
```

```python
#Finding the condition of cars using find_all
#All the condition are present in span class : certified
condition = []
a=soup.findAll('span',{"class":"certified"})
for i in a :
    print(i.text)#printing condition
    condition.append(i.text)#appending values
print(condition)#printing condition
len(condition)#printing length of list
```

```python
#Finding the price of cars using find_all
#All the price are present in div class : actual
car_price = []
a=soup.findAll('div',{"class":"actual"})
for i in a :
    print(i.text)#printing price
    car_price.append(i.text.replace("* Price",""))#appending values
print(car_price)#printing price
len(car_price)#printing length of list
```

```python
#Finding the stock of cars using find_all
#All the price are present in small class : stock
stock = []
a=soup.findAll('small',{"class":"stock"})
for i in a :
    print(i.text)#printing stock
    stock.append(i.text.replace("Stock: ",""))#appending values
print(stock)#printing stock
len(stock)#printing length of list
```

```python
#Finding the vin of cars using find_all
#All the vin are present in small class : vin
vin = []
a=soup.findAll('small',{"class":"vin"})
for i in a :
    print(i.text.strip())#printing vin
    vin.append(i.text.replace("VIN: ","").strip())#appending values
print(vin)#printing vin
len(vin)#printing length of list
```

```python
#creating dict of scrapped data
dict = {
    'Title' : car_titles,
    'Price' : car_price,
    'Condition' : condition,
    'Stock':stock,
    'VIN':vin
}
```

```python
df = pd.DataFrame.from_dict(dict,orient = 'index').T
df
```

```python
#Dataframe to csv
df.to_csv('funkstoyota.csv')
```
