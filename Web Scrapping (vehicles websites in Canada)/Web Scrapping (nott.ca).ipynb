```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from selenium import webdriver
import time
from matplotlib import pyplot as plt
```

```python
urla = input(("Enter url : "))
# "https://www.nott.ca/used-vehicles/"
```

```python
driver=webdriver.Firefox(executable_path="C:/Users/DELL/Downloads/geckodriver.exe")
#set the value of executable path
driver.get(urla)
```

```python
page_source=driver.page_source
```

```python
#parsing the page
soup=BeautifulSoup(page_source,'html.parser')
```

```python
#Finding the titles of cars using find_all
#All the titles are present in div class : col-xl-3 col-lg-4 col-md-6 col-sm-12 col-xs-12
a = soup.find_all("div",{"class" :"col-xl-3 col-lg-4 col-md-6 col-sm-12 col-xs-12"})
#taking an empty list
car_titles = []
#printing all the separate titles using for loop
for i in a:
    titles = i.find('div',{"class":"title"}).text #diggig down for titles
    print(titles)# printing values
    #appending values in a separate empty list
    car_titles.append(titles)  
print(car_titles)#printing car titles list
len(car_titles)#length of car titles list
```

```python
#All the mileage are present in table class : options-primary
b = soup.findAll("table",{"class" :"options-primary"})
#taking an empty list
car_mileage = []
#printing all the separate mileage using for loop
for i in b:
    x = i.find("tr",{"class" :"listing_category_mileage"}).text.replace("Kilometers: ","")#digging further to get mileage
    print(x.strip())
        #print(p)# printing values
    #appending values in a separate empty list
    car_mileage.append(x.strip())  
print(car_mileage)#printing car mileage list
len(car_mileage)#length of car mileage list
```

```python
#All the drivetrain are present in table class : options-primary
c = soup.findAll("table",{"class" :"options-primary"})
#taking an empty list
car_drivetrain = []
#printing all the separate drivetrain using for loop
for i in c:
    x = i.find("tr",{"class" :"listing_category_drivetrain"}).text.replace("Drivetrain: ","")#digging further to get drivetrain
    print(x.strip())# printing values
    #appending values in a separate empty list
    car_drivetrain.append(x.strip())  
print(car_drivetrain)#printing car drivetrain list
len(car_drivetrain)#length of car drivetrain list
```

```python
#All the bodystyle are present in table class : options-primary
c = soup.findAll("table",{"class" :"options-primary"})
#taking an empty list
car_bodystyle  = []
#printing all the separate bodystyle using for loop
for i in c:
    x = i.find("tr",{"class" :"listing_category_body-style"}).text.replace("Body Style: ","")#digging further to get bodystyle 
    print(x.strip())# printing values
    #appending values in a separate empty list
    car_bodystyle .append(x.strip())  
print(car_bodystyle )#printing car bodystyle  list
len(car_bodystyle )#length of car bodystyle  list
```

```python
#All the exterior are present in div class : options-primary
b = soup.findAll("table",{"class" :"options-primary"})
#taking an empty list
car_exterior = []
#printing all the separate exterior using for loop
for i in b:
    x = i.find("tr",{"class" :"listing_category_exterior-color"}).text.replace("Exterior Color: ","")#digging further to get exterior
    print(x.strip())# printing values
    #appending values in a separate empty list
    car_exterior.append(x.strip())  
print(car_exterior)#printing car exterior list
len(car_exterior)#length of car exterior list
```

```python
#All the interior are present in div class : options-primary
b = soup.findAll("table",{"class" :"options-primary"})
#taking an empty list
car_interior = []
#printing all the separate interior using for loop
for i in b:
    x = i.find("tr",{"class" :"listing_category_interior-color"}).text.replace("Interior Color: ","")#digging further to get interior
    print(x.strip())# printing values
    #appending values in a separate empty list
    car_interior.append(x.strip())  
print(car_interior)#printing car interior list
len(car_interior)#length of car interior list
```

```python
#Finding the price of cars using find_all
#All the price are present in div class : price
c = soup.findAll("div",{"class" :"price"})
#taking an empty list
car_price = []
#printing all the separate price using for loop
for i in c:
    prices = i.find("div",{"class" :"figure"}).text.strip()#digging further to get price
    print(prices)# printing values
    #appending values in a separate empty list
    car_price.append(prices)  
print(car_price)#printing car price list
len(car_price)#length of car price list
```

```python
#creating dict of scrapped data
dict = {
    'Title' : car_titles ,
    'Mileage' : car_mileage , 
    'Drive Train' : car_drivetrain ,  
    'Price ': car_price,
    'Interior' : car_interior,
    'Exterior' : car_exterior,
    'Body Style' :car_bodystyle,
    'Drive Train' :car_drivetrain
}
```

```python
#creating dataframe from dict
df = pd.DataFrame.from_dict(dict,orient = 'index').T
df
```

```python
#Dataframe to csv
df.to_csv('nott_ca.csv')
```
