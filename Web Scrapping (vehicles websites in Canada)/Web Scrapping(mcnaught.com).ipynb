```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from selenium import webdriver
import time
from matplotlib import pyplot as plt
```

```python
urla = "https://www.mcnaught.com/new-vehicles/"
```

```python
driver=webdriver.Firefox(executable_path="C:/Users/DELL/Downloads/geckodriver.exe")
#set the value of executable path
driver.get(urla)
```

```python
page_source=driver.page_source
```

```python
#parsing the page
soup=BeautifulSoup(page_source,'html.parser')
```

```python
#All the data present in a page is present in class: vehicle-cards
data = soup.findAll('div',{"class":"vehicle-cards"})
#printing the extracted data of a single page
for info in data:
    print(info.text)
```

```python
#Finding the make years of cars using find_all
#All the make years are present in h2 class : result-title
a = soup.find_all("h2",{"class" :"result-title"})
#taking an empty list
car_makeyear = []
#printing all the separate make years using for loop
for i in a:
    my = i.find('span',{"class":"title-top"}).text #diggig down for make years
    print(my)# printing values
    #appending values in a separate empty list
    car_makeyear.append(my)  
print(car_makeyear)#printing car make years list
len(car_makeyear)#length of car make years list
```

```python
#Finding the titles of cars using find_all
#All the titles are present in h2 class : result-title
a = soup.find_all("h2",{"class" :"result-title"})
#taking an empty list
car_title = []
#printing all the separate titles using for loop
for i in a:
    title = i.find('span',{"class":"title-bottom"}).text #digging down for titles
    print(title)# printing values
    #appending values in a separate empty list
    car_title.append(title)  
print(car_title)#printing car titles list
len(car_title)#length of car titles list
```

```python
#Finding the stocks of cars using find_all
#All the stocks are present in div class : hit-content-title-wrap
b = soup.find_all("div",{"class" :"hit-content-title-wrap"})
#taking an empty list
car_stocks = []
#printing all the separate stocks using for loop
for i in b:
    stocks = i.find('div',{"class":"stock-row"}).text #digging down for stocks
    print(stocks)# printing values
    #appending values in a separate empty list
    car_stocks.append(stocks.replace("Stock:",""))  
print(car_stocks)#printing car stocks list
len(car_stocks)#length of car stocks list
```

```python
#Finding the market value of cars using find_all
#All the market value are present in div class : hit-price hit-price--extra
c = soup.find_all("div",{"class" :"hit-price hit-price--extra"})
#taking an empty list
m_value = []
#printing all the separate market value using for loop
for i in c:
    mv = i.find('div',{"class":"hit-price__value"}).text #digging down for market value
    print(mv)# printing values
    #appending values in a separate empty list
    m_value.append(mv)  
print(m_value)#printing car market value list
len(m_value)#length of car market value list
```

```python
#Finding the final price of cars using find_all
#All the final price value are present in div class : hit-price hit-price--primary
d = soup.find_all("div",{"class" :"hit-price hit-price--primary"})
#taking an empty list
car_price = []
#printing all the separate final price using for loop
for i in d:
    price = i.find('div',{"class":"hit-price__value"}).text #diggig down for final price
    print(price)# printing values
    #appending values in a separate empty list
    car_price .append(price)  
print(car_price )#printing car final price list
len(car_price )#length of car final price list
```

```python
#All the exterior are present in ul class : vehicle-details
b = soup.findAll("ul",{"class" :"vehicle-details"})
#taking an empty list
car_exterior = []
#printing all the separate exterior using for loop
for i in b:
    x = i.find("li",{"class" :"vehicle-details--item vehicle-details--item-color exterior"}).text.replace("Exterior: ","")#digging further to get exterior
    print(x.strip())# printing values
    #appending values in a separate empty list
    car_exterior.append(x.strip())  
print(car_exterior)#printing car exterior list
len(car_exterior)#length of car exterior list
```

```python
#All the interior are present in ul class : vehicle-details
b = soup.findAll("ul",{"class" :"vehicle-details"})
#taking an empty list
car_exterior = []
#printing all the separate interior using for loop
for i in b:
    x = i.find("li",{"class" :"vehicle-details--item vehicle-details--item-color interior"}).text.replace("Interior: ","")#digging further to get exterior
    print(x.strip())# printing values
    #appending values in a separate empty list
    car_exterior.append(x.strip())  
print(car_exterior)#printing car interior list
len(car_exterior)#length of car interior list
```

```python
#creating dict of scrapped data
dict = {
    'Title' : car_title ,
    'Make Year' : car_makeyear,
    'Stock' : car_stocks, 
    'Market Value':m_value,  
    'Price ': car_price,
    'Car Exterior' : car_exterior,
}
```

```python
#creating dataframe from dict
df = pd.DataFrame.from_dict(dict,orient = 'index').T
df
```

```python
df['Dealer'] = 'mcnaught'
```

```python
split_make_year = df["Make Year"].str.split(" ", n = 1, expand = True) 
# splitting Tile column 
```

```python
df["Condition"]= split_make_year[0] 
```

```python
df["Make Year"]= split_make_year[1] 
```

```python
df["Model"]= split_title[1] 
```
