```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from selenium import webdriver
import time
from matplotlib import pyplot as plt
```

```python
urla = "https://ebrandon.ca/browseads.aspx?ward=&bus_name=&person_id=&seller_id=&feature=&wanted=&favourited=&issold=&notsold=&orderby=&tag=223,&expiring=&bargain=&free=&reduced=&item_business=&keywords=&cities=&cat_id=63&notcityid=&notsiteid=&notprovinceid=&pricelow=&pricehigh=&viewmode=list"
```

```python
driver=webdriver.Firefox(executable_path="C:/Users/DELL/Downloads/geckodriver.exe")
#set the value of executable path
driver.get(urla)
```

```python
page_source=driver.page_source
```

```python
#parsing the page
soup=BeautifulSoup(page_source,'html.parser')
```

```python
#All the data present in a page is present in class section_wrap classified_items_wrap
data=soup.findAll('div',{"class":"section_wrap classified_items_wrap"})
#printing the extracted data of a single page
for info in data:
    print(info.text)
```

```python
#Finding the titles of cars using find_all
#All the titles are present in div class : large-6 medium-7 columns
car_titles = []
a=soup.findAll('div',{"class":"large-6 medium-7 columns"})
for i in a :
    x = i.find('h6') # digging further
    print(x.text)# printing titles
    car_titles.append(x.text)# appending values
print(car_titles)# printing titles
len(car_titles)# printing length of list
```

```python
#Finding the location of cars using find_all
#All the location are present in span class : show-for-medium
car_location = []
a=soup.findAll('span',{"class":"show-for-medium"})
for loc in a :
    print(loc.text.strip())#printing location
    car_location.append(loc.text.strip())# appending values
print(car_location)#printing location
len(car_location)#printing length of list
```

```python
#Finding the car price using find_all
#All the car price are present in div class : large-3 medium-5 columns hide_content_on_grid
car_price = []
a=soup.findAll('div',{"class":"large-3 medium-5 columns hide_content_on_grid"})
for c in a :
    x = c.find('span',{"class":"price"}) #digging further
    print(x.text.strip())#printing car price
    car_price.append(x.text.strip())#appending values
print(car_price)#printing car price
len(car_price)#printing length of list
```

```python
#creating dict of scrapped data
dict = {
    'Title' : car_titles ,
    'Location':car_location,
    'Price ': car_price 
}
```

```python
#creating dataframe from dict
df = pd.DataFrame.from_dict(dict,orient = 'index').T
df
```

```python
df['Condition'] = 'Used'
df['Dealer'] = 'ebrandon'
```

```python
split_title = df["Title"].str.split(" ", n = 2, expand = True) 
# splitting Tile column 
df["Make Year"]= split_title[0] 
df["Make"]= split_title[1] 
df["Model"]= split_title[2] 
df
```

```python
df.drop(columns =["Title"], inplace = True)
df
```

```python
#Dataframe to csv
df.to_csv('ebandron.ca.csv')
```
