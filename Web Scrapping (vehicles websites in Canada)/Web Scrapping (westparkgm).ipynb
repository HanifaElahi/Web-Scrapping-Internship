```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from selenium import webdriver
import time
from matplotlib import pyplot as plt
```

```python
urla = "https://www.westparkgm.com/VehicleSearchResults"
```

```python
driver=webdriver.Firefox(executable_path="C:/Users/DELL/Downloads/geckodriver.exe")
#set the value of executable path
driver.get(urla)
```

```python
page_source=driver.page_source
```

```python
#parsing the page
soup=BeautifulSoup(page_source,'html.parser')
```

```python
#All the data present in a page is present in class deck
data=soup.findAll('div',{"class":"deck"})
#printing the extracted data of a single page
for info in data:
    print(info.text)
```

```python
#Finding the titles of cars using find_all
#All the titles are present in h4 template : vehicle-name
car_titles = []
a=soup.findAll('h4',{"template":"vehicle-name"})
for i in a :
    print(i.text)#printing titles
    car_titles.append(i.text)#appending values
print(car_titles)#printing titles
len(car_titles)#printing length of list
```

```python
interior = []
engine = []
exterior = []
cylinders = []
vin = []
stock = []
# taking empty lists

spec= soup.findAll('ul',{"class":"vehicleIdentitySpecs"}) #all the specifications are inside class  vehicleIdentitySpecs
for i in spec:
    newsoup = BeautifulSoup(str(i),'html.parser')#using parser for scrapping inside details
    lis = newsoup.find_all('li')#finding all li
    for li in lis:
        
        if 'Engine' in li.text: #if present
            engine.append(li.text.replace("Engine\n\n\n","").strip())#then append
        if 'Interior' in li.text:#if present
            interior.append(li.text.replace("Interior\n","").strip())#then append
        if 'Cylinders' in li.text:#if present
            cylinders.append(li.text.replace("Cylinders\n","").strip())#then append
        if 'Exterior' in li.text:#if present
            exterior.append(li.text.replace("Exterior\n","").strip())#then append
        if 'VIN' in li.text:#if present
            vin.append(li.text.replace("VIN\n\n","").strip()) #then append            
        if 'Stock Number' in li.text:#if present
            stock.append(li.text.replace("Stock Number\n","").strip())#then append
#printing lists            
print(engine)
print(interior)
print(cylinders)
print(exterior)
print(vin)
print(stock)
```

```python
msrp= []
car_price = []
savings = []
# taking empty lists

spec= soup.findAll('ul',{"class":"prices"})  #main class
for i in spec:
    newsoup = BeautifulSoup(str(i),'html.parser')#using parser for scrapping inside details
    lis = newsoup.find_all('li')#finding all li
    for li in lis:
        
        if 'Your Price' in li.text:
            car_price.append(li.text.replace("Your Price\n\n","").strip())
        
        
        if 'Savings' in li.text:
            savings.append(li.text.replace("Savings\n","").strip())

print(car_price)
print(savings)


```

```python
#creating dict of scrapped data
dict = {
    'Title' : car_titles,
    'Engine' : engine ,
    'Cylinders':cylinders,
    'Interior':interior,
    'Exterior':exterior,
    'VIN':vin,
    'Stock':stock,
    'Price' : car_price,
    'Savings':savings
}
```

```python
df = pd.DataFrame.from_dict(dict,orient = 'index').T
df
```

```python
df['Dealer'] = 'westparkgm'
```

```python
split_title = df["Title"].str.split("\n", n = 7, expand = True) 
# splitting Tile column 
df["Condition"]= split_title[1] 
df["Make Year"]= split_title[2] 
df["Make"]= split_title[5] 
df["Model"]= split_title[6] 
df
```

```python
df.drop(columns =["Title"], inplace = True)
df
```

```python
#Dataframe to csv
df.to_csv('westparkgm.csv')
```
