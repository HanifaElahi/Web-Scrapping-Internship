```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from selenium import webdriver
import time
from matplotlib import pyplot as plt
```

```python
urla = "https://www.virdenmainline.com/new-vehicles/"
```

```python
driver=webdriver.Firefox(executable_path="C:/Users/DELL/Downloads/geckodriver.exe")
#set the value of executable path
driver.get(urla)
```

```python
page_source=driver.page_source
```

```python
#parsing the page
soup=BeautifulSoup(page_source,'html.parser')
```

```python
#All the data present in a page is present in id lvrp-results-container
data=soup.findAll('div',{"id":"lvrp-results-container"})
#printing the extracted data of a single page
for info in data:
    print(info.text)
```

```python
condition = []
a=soup.findAll('span',{"class":"title-top"})
for i in a :
    print(i.text)#printing condition
    condition.append(i.text)#appending values
print(condition)#printing condition
len(condition)#printing length of list
```

```python
#Finding the titles of cars using find_all
#All the titles are present in h2 class : result-title
car_titles = []
a=soup.findAll('span',{"class":"title-bottom"})
for i in a :
    print(i.text)#printing titles
    car_titles.append(i.text)#appending values
print(car_titles)#printing titles
len(car_titles)#printing length of list
```

```python
exterior = []
interior = []
# taking empty lists

spec= soup.findAll('ul',{"class":"vehicle-details"}) # all the details are inside ul class : vehicle-details

for i in spec:
    newsoup = BeautifulSoup(str(i),'html.parser')#using parser for scrapping inside details
    lis = newsoup.find_all('li')#finding all li inside ul
    for li in lis:
        
        if 'Exterior' in li.text: #if present
            exterior.append(li.text.replace("Exterior:",""))#then append
        if 'Interior' in li.text:#if present
            interior.append(li.text.replace("Interior:",""))#then append
       
        
#printing individual lists        
print(exterior)
print(interior)
```

```python
car_price = []#taking an empty list
a = soup.findAll('div',{"class":"price-block our-price real-price"})#scrapping price
for i in a:
    print(i.text.strip())
    car_price.append(i.text.replace("MSRP","").strip())#appending in a list
print(car_price)#printing list
```

```python
#creating dict of scrapped data
dict = {
    'Title' : car_titles ,
    'Exterior' : exterior , 
    'Interior': interior,
    'Price' : car_price,
    'Condition':condition
}
```

```python
df = pd.DataFrame.from_dict(dict,orient = 'index').T
df
```

```python
df['Dealer'] = 'vidermainline'
```

```python
split_title = df["Title"].str.split(" ", n = 3, expand = True) 
# splitting Tile column 
df["Make Year"]= split_title[0] 
df["Make"]= split_title[1] 
df['Model'] = split_title[2]

split_condition = df["Condition"].str.split(" ", n = 1, expand = True) 
# splitting Tile column 
df["Condition"]= split_condition[0] 
df
```

```python
df.drop(columns =["Title"], inplace = True)
df
```

```python
#Dataframe to csv
df.to_csv('vidermainline.csv')
```
