```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from selenium import webdriver
import time
from matplotlib import pyplot as plt
```

```python
urla = "https://www.carmanford.ca/vehicles/?v1=car&st=price,asc&view=grid"
```

```python
driver=webdriver.Firefox(executable_path="C:/Users/DELL/Downloads/geckodriver.exe")
#set the value of executable path
driver.get(urla)
```

```python
page_source=driver.page_source
```

```python
#parsing the page
soup=BeautifulSoup(page_source,'html.parser')
```

```python
#All the data present in a page is present in div class : row grid-row
data=soup.findAll('div',{"class":"row grid-row"})
#printing the extracted data of a single page
for info in data:
    print(info.text)
```

```python
#Finding the titles of cars using find_all
#All the titles are present in div class : vehicle-card__full-details
car_titles = []
a=soup.findAll('div',{"class":"vehicle-card__full-details"})
for i in a :
    titles=i.find('h3',{"class":"vehicle-card__title mb-xs"}).text.strip()#all the titles are inside h3 class
    print(titles)#printing titles
    car_titles.append(titles)#appending values
print(car_titles)#printing titles
len(car_titles)#printing length of list
```

```python
#Finding the location of cars using find_all
#All the location are present in p class : vehicle-card__location
car_location = []
b = soup.findAll('p',{"class":"vehicle-card__location"})
for i in b :
    print(i.text)
    car_location.append(i.text.replace("Located at ",""))#apppending inside a separate list
print(car_location)#printing list
len(car_location)#length of list
```

```python
vin = []
mileage = []
condition = []
body_style = []
stock = []
# taking empty lists

spec= soup.find_all("li",{"class":"detailed-specs__single"}) #all the specifications are inside li class: detailed-specs__single

for i in spec:
    
    if 'VIN' in i.text: #if present
        vin.append(i.text.replace("VIN: ",""))#then append
    if 'Kilometres' in i.text:#if present
        mileage.append(i.text.replace("Kilometres: ",""))#then append
    if 'Condition' in i.text:#if present
        condition.append(i.text.replace("Condition: ",""))#then append
    if 'Body Style' in i.text:#if present
        body_style.append(i.text.replace("Body Style: ",""))#then append
    if 'Stock' in i.text:#if present
        stock.append(i.text.replace("Stock #: ",""))#then append
        
#printing individual lists        
print(vin)
print(mileage)
print(condition)
print(body_style)
print(stock)
```

```python
#Finding the msrp of cars using find_all
#All the msrp are present in div class : price-block
msrp = []
c = soup.findAll('div',{"class":"price-block"})
for i in c :
    sp=i.findAll('div',{"class":"price-block__price price-block__price--md price-block__price--grey"})#digging further
    for j in sp:
        s=j.find('span',{"class":"df aifs"}).text.split("df aifs")[0]#actual values are inside span class :df aifs
        print(s)#printing msrp
        msrp.append(s)#appending in a list
print(msrp)#printing list
len(msrp)#length of list
```

```python
#Finding the savings on cars using find_all
#All the  savings  are present in div class : vehicle-card__price-wrap
savings = []
c = soup.findAll('div',{"class":"vehicle-card__price-wrap"})
for i in c :
    sp=i.findAll('div',{"class":"price-block__price price-block__price--sm price-block__price--primary"})#digging further
    for j in sp:
        s=j.find('span',{"class":"df aifs"}).text.split("df aifs")[0]#actual values are inside span class :df aifs
        print(s)#printing initial prices
        savings.append(s)#appending in a list
print(savings)#printing list
len(savings)#length of list
```

```python
#Finding the sales price on cars using find_all
#All the  sales price  are present in div class : price-block__single price-block--expanded
car_price = []
c = soup.findAll('div',{"class":"price-block__single price-block--expanded"})
for i in c :
    sp=i.findAll('div',{"class":"price-block__price price-block__price--lg price-block__price--dark"})#digging further
    for j in sp:
        s=j.find('span',{"class":"df aifs"}).text.split("df aifs")[0]#actual values are inside span class :df aifs
        print(s)#printing prices
        car_price.append(s)#appending in a list
print(car_price)#printing list
len(car_price)#length of list


```

```python
#creating dict of scrapped data
dict = {
    'Title' : car_titles ,
    'Location' : car_location , 
    'Mileage ': mileage ,
    'VIN' : vin ,
    'Stock' : stock ,
    'Condition ': condition,
    'Body Style' : body_style ,
    'Stock' : stock,
    'MSRP' : msrp,
    'Savings' : savings,
    'Price' : car_price
}
```

```python
df = pd.DataFrame.from_dict(dict,orient = 'index').T
df
```

```python
split_title = df["Title"].str.split(" ", n = 3, expand = True) 
# splitting Tile column 
```

```python
df["Make Year"]= split_title[0]  
```

```python
df["Make"]= split_title[1]
```

```python
df["Model"]= split_title[2]
```

```python
df.drop(columns =["Title"], inplace = True)
```
